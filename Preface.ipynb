<h2> はじめに</h2>
</div>

&emsp;本修士論文では,&thinsp; 具体的な多様体上にある点データ群（データクラウド）に対するパーシステントホモロジーを定義し,&thinsp; 人工データによる計算機実験を行なった.&thinsp; また,&thinsp; 多様体の向きづけ可能性などに関わるコホモロジー特性類のパーシステントホモロジー版の導出と検証について考察した.  
&emsp;画像認識などのデータサイエンスの分野において最近傍法やニューラルネットなどさまざまな方法が提案されているが,&thinsp; 近年,&thinsp; 位相的データ解析（Topological Data Analysis）を総称される位相幾何学的分析手法が注目されている.&thinsp; とくにパーシステントホモロジー群は,&thinsp; データの集合が全体的にどのような位相的特徴をもっているのかを表現するのに優れた道具である. &thinsp;実際に,&thinsp; タンパク質の折りたたみ構造の分析,&thinsp; 重みがついたネットワークのトポロジーの分析,&thinsp; ガラスなどのアモルファス素材の構造解析,&thinsp; ゲノム解析などに使用されて成果を上げている.  
&emsp;さて,&thinsp; いわゆる多様体仮説とは,&thinsp; 高次元ユークリッド空間上にスパースに分布する点データ群が低次元のある部分多様体の近傍に一様に分布しているだろうとする仮説である.&thinsp; この多様体仮説の具体的な一例として,&thinsp; Carlsson,&thinsp; Mumfordらによる自然背景画像の位相的データ解析の研究がある[1,&thinsp; 2,&thinsp; 3].&thinsp; 数百万個のグレースケールの自然画像データから$3 \times 3$パッチをランダムに取ったところ,&thinsp; パッチのデータ空間上に点データ群がスパースに分布し,&thinsp; パーシステントホモロジーの計算を通じて,&thinsp; ある2次元部分多様体（閉曲面）の近くに高密度にデータが存在することが検証された.&thinsp; とくにこの多様体はクライン・ボトルに同相であって向きづけ不可能な曲面として知られている.  
&emsp; そこで本修士論文では,&thinsp; Mumford らの研究を動機づけとして,&thinsp; 逆に多様体$M$とその上に分布する点データ群$D$が与えられたとき,&thinsp; $D$のパーシステントホモロジーをどのように定義するかという問いから考察をはじめた.&thinsp; まずあらかじめ多様体$M$に距離を導入して点データ群$D$の各$2$点間の距離を計算し,&thinsp; 隣接行列,&thinsp; すなわち重みつきグラフを生成する.&thinsp; このグラフは仮想的に高次元のユークリッド空間に埋め込まれるだろう.&thinsp; これはいわばナッシュの埋め込み定理で保証されるリーマン多様体$M$のユークリッド空間への埋め込みの仮想的な代替ともみなせる.&thinsp; このユークリッド空間内の点データ群のヴィートリス・リップス複体によるパーシステントホモロジー群を考える.&thinsp; これは多様体$M$のホモロジーを近似するものではないが,&thinsp; 点データ群$D$がある程度$M$上で一様に分布しているならば,&thinsp; ある安定した時刻帯において$M$のホモロジーを復元すると考えられる.&thinsp; Mumfordらの研究に関連して,&thinsp; $M$を閉曲面,&thinsp; 特にクライン・ボトルやトーラスを扱うことにして,&thinsp; 人工的に生成した点データ群を用いて計算機実験を行った.&thinsp; また,&thinsp; 一般的な議論として$M$上の増大するフィルトレーションが与えられたとき,&thinsp; $M$の接束$TM$のホイットニー類のパーシステント版を定義した.&thinsp; さらに,&thinsp; 上述の計算機実験に現れるバーコードにおいて,&thinsp; パーシステントホイットニー類を同定した.&thinsp; 将来的には&thinsp; 高次元のトーラスや球面上の方向統計学における実データを用いて,&thinsp; 本修論で考察した手法を適用し,&thinsp; 検証したいと考える. &thinsp;
