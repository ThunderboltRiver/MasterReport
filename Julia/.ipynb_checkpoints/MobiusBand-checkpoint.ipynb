実射影空間$RP^1$上の点群に対するパーシステントホモロジー

閉区間$I = [0, 1]$上の点を実射影空間の点に変換する方法  
$I$に同値関係$\sim$を$0 \sim 1, q \sim q  (q\in [0, 1])$を定義することによって$I / \sim$は$S^1$に同相になる.その同相写像は$f([x]) = (\cos(2\pi x), \sin(2\pi x))$で与えられる.$RP^1:= \{ \{x, -x\} | x \in S^1 \}$より,角度が$\pi \leq \theta < 2 \pi$の点は$ \theta - \pi$に同一視される.従って$I$の部分集合$P$を$RP^1$上の点群へ変換するには$P' := (P \cap [0, 1/2) ) \cup \{ x - 1 / 2 | x \in P \cap [1/2, 1)\}$
