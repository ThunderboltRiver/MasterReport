# 1.ホモロジー
位相幾何学では,球体やトーラスなどのユークリッド空間内の図形から連結成分や輪っかや空洞の数と言った幾何学的な特徴を捉える際に,一度図形を簡単な多面体に近似(変形)して,代わりにこの多面体の特徴を捉えるといった手法を用いる.しかし雑に近似したところで,元の図形の関心のある特徴が大きく崩れては意味がない.そこで用いられる主な近似の手法として,位相同型やホモトピー同値が挙げられる.これら二つの変形によって,図形の見た目を大きく変えたとしても関心のある特徴を保存することができる.例えば,連結成分や輪っかや空洞の数について関心があれば元の図形とホモトピー同値な多面体を調べれば良い.それよりももっと多くのことを調べる際は,ホモトピー同値よりも近似度合いの強い位相同型な多面体について調べれば良い.このようにして考察の対象を多面体のみに狭めることができる.\
そこでこのセクションでは多面体とは何かについて述べた後,連結成分や輪っかや空洞の数を調べる際の代数学的な概念であるホモロジー群について解説する.その後,本論文での考察対象となるホモロジー多様体について解説する.

## 1.1多面体
多面体は,単体と呼ばれる簡単な図形がいくつか合併したものと捉え,その単体の集まりが単体複体として定義される.そこで先に単体と単体複体について解説する.

### 定義1.1.1 単体
$R^N$内の$n + 1$個の点$a_0, \cdots , a_n$に対して,ベクトル$\vec{a_0a_1}, \cdots, \vec{a_0a_n}$が一次独立であるとき,集合
$$|a_0, \cdots, a_n| = \{\sum_{i = 0}^{n} \lambda_i a_i | \lambda_i \geq 0 (i = 0, \cdots, n) \sum_{i = 0}^{n}\lambda_i = 1\}$$
を,$a_0, \cdots, a_n$を頂点とする **$n$次元単体** といい,$\sigma = |a_0, \cdots, a_n|$などと書く.\
$\sigma$が単体であるとき,$\sigma$の頂点集合$\{a_0, \cdots, a_n\}$の空でない部分集合を頂点とする単体$\tau$を$\sigma$の**辺単体**といい,$\tau \preceq \sigma$と書く.単体$\sigma$が$n$次元であることを$\dim \sigma = n$と書く.

### 定義1.1.2 単体複体と多面体
$R^N$内のいくつかの単体からなる有限集合$K$であって,以下の条件を満たすものを **単体複体** という.
$$(i) \sigma \in K \tau \preceq \sigma \Longrightarrow \tau \in K$$
$$(ii) \sigma, \tau \in K \sigma \cap \tau \ne \emptyset \Longrightarrow  \sigma \cap \tau \preceq \sigma, \tau$$
$K$の$0$単体は **$K$の頂点** とよばれ,それら全体の集合を$\hat K$などと書く. $n:= \max{\{\dim \sigma | \sigma \in K \}}$を単体複体 $K$ の**次元**といい,$\dim K = n$と書く.
また,単体複体$K$の全ての単体の和集合
$$|K| :=  \bigcup_{\sigma \in K} \sigma$$
を$K$の**多面体**という.

### 細分と重心細分

ホモロジー多様体や後のセクションで述べるパーシステント性において必要になる部分複体とその具体例について述べる.
### 定義1.1.3 部分複体
単体複体$K$の部分集合$L \subset K$がまた単体複体であるとき,$L$を$K$の **部分複体** という.\
$L \subset K$が部分複体であるための必要十分条件は,定義1.1.2の条件$(i)$を満たすことである.

### 例1.1.4
$K$を単体複体とし,$p \in |K|$を多面体の点とする.
$$S_K(p) := \{\tau \in K | \exists \sigma \in K s.t. \ p \in \sigma , \tau \prec \sigma \}$$
$$L_K(p):= \{ \tau \in S_K(p)| p \notin \tau \}$$
とおくと,これらはともに$K$の部分複体である.それぞれ **$K$における$p$の星状複体,まつわり複体** という.
&emsp;直感的な言い方をすると,$p \in |K|$のまつわり複体とは,多面体$|K|$内で$p$に触れることなく囲うように位置する$K$の単体を全て集めた単体複体といえる.

&emsp;多面体を構成するには単体複体を構成する必要があるが,単体はユークリッド空間の図形であるからいちいちこれを作り上げるのは手間がかかる.また,多面体の辺の長さや面積,体積,内角といった量に元から関心がないなら,一般の位置にある点を具体的に考える必要はない.\
&emsp;例えば多面体として三角形を何か一つ構成する際,どんな形だろうととにかく三角形でありさえすれば良い場合は具体的な頂点を取る必要がなく,あくまで$3$つの頂点が一般的な位置にあるという情報だけで事足りる.\
&emsp;この視点で単体複体の頂点全体に注目すると,各単体はその頂点の組み合わせ(部分集合)として捉えられることがわかる.より具体的には次の命題1.1.5が成り立つ.

### 命題1.1.5
$K$を単体複体とする.$K$の頂点全体の集合$\hat K$の部分集合族$\Sigma$を,$K$の単体の頂点集合全体として定義する.
$$\Sigma := \{\{a_0, \cdots, a_k \} \subset \hat K | k \geq 0, |a_0, \cdots, a_k| \in K \} \cup \{ \emptyset \}$$
このとき$\hat K$と$\Sigma$は次のような性質を持つ.
$$(i) \forall v \in \hat K, \{v\} \in \Sigma $$
$$(ii)\hat \sigma \in \Sigma, \hat \tau \subset \hat \sigma \Longrightarrow \hat \tau \in \Sigma $$
$proof)$
$(i):v \in \hat K$は$K$の頂点,つまり$0$単体であるから$|v| = v \in K$である.従って$\{ v \} \in \Sigma$\
$(ii):\hat \sigma \in \Sigma$を頂点集合とする$K$の単体を$|\hat \sigma|$とする.$\hat \tau \subset \hat \sigma$が空でないなら,$|\hat \tau| \prec |\hat \sigma|$である.$K$が複体であることから,定義1.1.2の条件$(i)$より$|\hat \tau| \in K$よって$\hat \tau \in \Sigma$ .$\square$

&emsp;命題1.1.5は幾何学的な概念である単体複体から条件$(i), (ii)$にあるような組み合わせ論的な構造を抽出できることを示している.そこで次のように定義する.
### 定義1.1.6 抽象複体
有限集合$V$とその部分集合族$\mathcal{S}$が以下の条件を満たすとき,対$(V, \mathcal{S})$を**抽象複体**と言う.
$$(i) \forall v \in V, \{ v \} \in \mathcal{S}$$
$$(ii) \hat \sigma \in \mathcal{S}, \hat \tau \subset \hat \sigma \Longrightarrow \hat \tau \in \mathcal{S} $$
このとき$V$は **頂点集合** と呼ばれ,その元を **頂点** という.$\mathcal{S}$ の元$\hat \sigma$は,その元の個数が$n + 1$であるなら $n$**次元単体** と呼ばれ,$\dim \hat \sigma = n$と書く.


&emsp;抽象複体という視点に立つと$n$単体は$n + 1$個の元からなる集合として捉えられ,それらの集合族によって単体複体の大まかな形が実現されていると捉えることができる.\
次に二つの抽象複体の間に定義される写像であって,単体を単体へ対応させるものについて定義する.
### 定義1.1.7 単体写像
二つの抽象複体$(V, \mathcal{S}), (V', \mathcal{S}' )$の頂点集合の間の写像$\phi: V \longrightarrow V' $は,以下の条件を満たすとき単体写像という.
$$\hat \sigma \in \mathcal{S} \Longrightarrow \phi(\hat \sigma) \in \mathcal{S}'$$
このとき$\phi: (V, \mathcal{S}) \longrightarrow (V', \mathcal{S}')$などと書く.また,二つの単体複体$KL$に命題1.1.5を適用することで得られる抽象複体の間に単体写像を構成したときは単に$\phi:K \longrightarrow L$などと書く.\
単体写像は単射であるとは限らないから,写った先の単体の次元は,元の単体より少ない可能性がある.もし単射であるなら次元は保たれる.\
この単体写像を用いて抽象複体の同型について定義する.これは幾何学的は,二つの単体複体に同じ組み合わせ的構造があることを示すものである.

### 定義1.1.8 同型な抽象複体
単体写像$\phi: (V, \mathcal{S}) \longrightarrow (V', \mathcal{S}')$は,全単射であり,その逆写像もまた単体写像であるとき,**単体同型写像** という.
さらにこのとき,二つの抽象複体$(V, \mathcal{S}),(V', \mathcal{S}')$は**同型**であるという.

### 組み合わせ同値

命題1.1.5は幾何学的な単体複体から抽象複体を構成したが,この逆もまた成り立つ.
### 命題1.1.9
抽象複体$(V, \mathcal{S})$に対して,命題1.1.5の$(\hat K, \Sigma)$と同型になるような単体複体$K$を構成できる.
この$K$を抽象複体$(V, \mathcal{S})$の**幾何学的実現**という.\
$proof)V = \{v_0, \cdots, v_N \}$とするとき,$\mathbb{R}^N$の標準基底$\boldsymbol{e}_i (i = 1, \cdots, N)$を取る.$\boldsymbol{e}_0 = \boldsymbol{0}$とするとき,$\boldsymbol{e}_0, \cdots, \boldsymbol{e}_N$は一般の位置にある.そこで単体複体$K$を次で定義する.
$$K := \{|\boldsymbol{e}_{i_0}, \cdots, \boldsymbol{e}_{i_k}| \ | \{v_{i_0}, \cdots, v_{i_k} \} \in \mathcal{S},  \{i_0, \cdots, i_k\} \subset \{0, \cdots, N\}, k \geq 0 \} \}$$
この$K$に命題1.1.5を適用して得られた$(\hat K, \Sigma)$と抽象複体$(V, \mathcal{S})$との間に写像$\phi:V \longrightarrow \hat K$を$\phi(v_i) = \boldsymbol{e}_i$で定義するとこれが同型な単体写像となる.$\square$

&emsp; 命題1.1.5と命題1.1.9により,単体複体を構成する際は抽象複体のみを考えるだけで十分であることがわかる.

&emsp;以上で多面体の解説を終えるが,最後に多面体を一般の図形に近似するための単体分割とホモトピー同値について述べる.
### 定義1.1.10 位相同型
2つの位相空間$X, Y$が位相同型であるとは,2つの連続写像$f:X \longrightarrow Y, g:Y \longrightarrow X$が存在して
$$g \circ f = id_{X} \ \ \ f \circ g = id_{Y}$$
となることをいう.ここで$id_{X}$は位相空間$X$の恒等写像である.

### 定義1.1.11 ホモトピック
2つの位相空間$X,Y$の間の連続写像の族$\{f_t\}_{t \in [0, 1]}:X \longrightarrow Y$を考える.このとき,写像
$$H(s, t) = f_t(s)$$
が連続であるとき,これを$f_0$と$f_1$の間のホモトピーという.またこのとき$f_0$と$f_1$はホモトピックであるといい$f_0 \simeq f_1$と書く.

### 定義1.1.12 ホモトピー同値
2つの位相空間$X,Y$の間に2つの連続写像$f: X \longrightarrow Y, g:Y \longrightarrow X$が存在して
$$g \circ f \simeq id_{X} \ \ \ f \circ g \simeq id_{Y}$$
となるとき,$X$と$Y$はホモトピー同値であるという.
### 定義1.1.11 単体分割
位相空間$M$が,ある単体複体$K$の多面体と位相同型であるとき,同相写像$t:K \longrightarrow M$を**単体分割**といい,$M$は単体分割可能であるという.

単体分割によって一般の図形を単体複体に置き換えて考察することが可能になる.


## 1.2 ホモロジー群
多面体内に$n$次元の穴がいくつあるかと言ったことを代数学的に調べる手法としてホモロジー群がある.ここでは一旦,ホモロジー群の直感的な説明を簡単に行う.\
多面体に含まれる図形は,単体複体内のいくつかの単体の和集合として大雑把に表現できる.従って,多面体内の穴もまたいくつかの単体の和集合として解釈できる.しかし,穴の情報だけが欲しいのなら,ループのような形をした図形のみを考えればよく,さらに言えば,「中身(内部)が埋まっているなら無視する」という規則を設けた方が便利である.\
上記の考え方は代数学的に厳密に表現することが可能で,特に「中身(内部)が埋まっているなら無視する」という規則は商群によって表現することができる.この商群をホモロジー群という.\
以下の定義1.2.1から1.2.?まではこれをもう少し厳密な言い方で表現する.そのためにまずは集合論的な和集合を,代数学的な形式和として捉え直すことから始める.これが鎖群の考え方である.


### 定義1.2.1 鎖群
$K$を単体複体とする.$K$の$q$次元単体全体からなる集合を基底とする,体$\mathbb{Z}_2$上のベクトル空間を$K$の$q$次元鎖群といい$C_q(K)$と書く.つまり,$\sigma_1^{q}, \cdots, \sigma_{k_q}^q$を$K$の$q$次元単体全体とするとき
$$C_q(K) := \{\sum_{i = 1}^{k_q}m_i \sigma_i^{q}| m_i \in \mathbb{Z}_2 \ , i = 1, \cdots, k_q\}$$
と定義する.$q > \dim K$と$q < 0$の時は$C_q(K) = \boldsymbol{0}$と定義する.\
ここで$c \in C_q(K)$を$q$次元鎖という.\
&emsp;鎖群の構成によって,集合論的な単体の和集合は鎖として表現される.ただし,鎖として単体を表すときは$\sigma = \langle a_0, \cdots, a_q \rangle$などと書くことにする.


次に「ループのような図形」や「中身が詰まっている」という表現をより詳細化するために境界準同型という鎖群間の写像を定義する.
### 定義1.2.2 境界準同型
整数$q$を$1 \leq q \leq \dim K$とするとき,$K$の$q$次元単体$\sigma =  \langle a_0, \cdots, a_q \rangle \in C_q(K)$の境界$\partial_q (\sigma) \in C_{q - 1}(K)$を以下で定義する.
$$\partial_q (\sigma) := \sum_{i = 0}^{q}\langle a_0, \cdots,a_{i - 1}, a_{i + 1}, \cdots a_q \rangle$$
さらに一般の$q$次元鎖$c = \sum_{i = 1}^{k_q} m_{i} \sigma_i^q \in C_q(K)$に対しては
$$\partial_q(c) := \sum_{i = 1}^{k_q}m_i \partial_q (\sigma_i^{q})$$
と定義する.ただし,ここで$\sigma_i^{q} (i =1, \cdots, k_q )$は単体複体$K$の$q$次元単体全体とする.\
$q > \dim K$あるいは$q \leq 0$となる$q$に対しては
$$\partial_q(c) := \boldsymbol{0} \ \ c \in C_q(K)$$
と定義する.
これによって任意の$q \in \mathbb{Z}$に対して準同型写像
$$\partial_q : C_q(K) \longrightarrow C_{q - 1}(K)$$
が定義される.これを **$q$次元境界準同型写像** という.\
　以上の準備のもとで「ループのような図形」をより厳密に「どの点からも始点と終点が同じであるように一筆書きできる図形」として定義し,その性質を考察することにする.\
　議論を簡単にするために一旦具体的に円周$S^1$についてのみ考える.まず$S^1$上に異なる３点をとることで単体分割が得られるが,そのどの頂点もちょうど2つの辺の頂点となっていることがわかる.これは点の数をどれだけ増やしていっても観測される.一方で,別の$S^1$と何処かの一点で同一視してできた図形(8の字ような図形)はどうなるだろうか.この場合,同一視された１点と他の異なる2点を,2つの$S^1$上からそれぞれ取ることで単体分割が得られ,そのどの頂点も偶数本の辺の頂点となっていることがわかる.\
　以上の考察から,位相空間$M$の内のループのような図形は,単体分割$t:K \longrightarrow M$とその鎖群$C_q(K)$の表現を用いると,次の条件を満たす鎖$c \in C_1(K)$として一般化できる.
$$\partial_1(c) = \boldsymbol{0}$$
そこで上記の性質を持つ鎖を**単体複体$K$の$1$次のサイクル**と呼ぶことにする.こうして「ループのような図形」は1次のサイクルとして厳密に表現できた.\
　しかし,一口に$1$次のサイクルといっても「中身が詰まっている」ものと「穴になっている」ものがある.たとえば円板$D^2$は$S^1$を含むから$1$次のサイクルを持っているが,同時に「中身が詰まっている」図形でもある.一方で$S^1$自身は「穴になっている」図形である.よって次にこれを区別する方法を考える.\
　簡単のため$D^2$に対して考察する.$D^2$の境界上に異なる点を３つ取ると,$D^2$はあたかも$2$単体を伸ばしてできた図形として見做せる.つまり$D^2$の単体分割として,$2$単体の辺単体全体が考えられる.この単体分割において$2$単体に境界準同型$\partial_2$を施すことで$D^2$の部分集合としての$S^1$を表す鎖が得られる.\
　以上の考察から,中身が詰まっている鎖$c \in C_1(K)$とは,
$$\exists c' \in C_2(K) \ s.t. \ \partial_2(c') = c$$
となるものであると解釈できる.このような鎖を**単体複体$K$の$1$次のバウンダリ**と呼ぶことにする.上記の例では,$D^2$の部分集合としての$S^1$を表す鎖が$1$次のバウンダリであるが,特に$1$次のサイクルでもある.これは偶然のように思われるかもしれないが,一般に成り立つことである.このことを次の命題に示す.



### 命題1.2.3
境界準同型$\partial_q:C_q(K) \longrightarrow C_{q - 1}(K)$と任意の$q$に対して
$$\partial_{q} \circ \partial_{q + 1} = \boldsymbol{0}$$
が成り立つ.\
$proof)q > \dim K$と$q \leq 0$のときは明らかだから$0 \leq q \leq \dim K$のときを示す.$\sigma = \langle a_0, \cdots, a_{q+1} \rangle$を$K$の単体とする.準同型性から$\partial_{q} (\partial_{q + 1} (\sigma)) = \boldsymbol{0}$を示せば良い.
$$\partial_q(\partial_{q +1}(\sigma)) = \partial_q(\sum_{i = 0}^{q + 1}\langle a_0, \cdots, \hat a_{i}, \cdots, a_{q + 1} \rangle) \\ = \sum_{i = 0}^{q + 1}\sum_{j< i}\langle a_0, \cdots, \hat a_{j}, \cdots, \hat a_{i}, \cdots, a_{q + 1} \rangle + \sum_{i = 0}^{q +1} \sum_{i< j}\langle a_0, \cdots, \hat a_{i}, \cdots, \hat a_{j}, \cdots, a_{q + 1} \rangle$$
ここで$\hat a_{i}$記号はその頂点を取り除くことを意味するものとする.$k < l$とすると上記の第一項において$i = l, j = l$としたものと第二項において$i = k, j = l$としたものは等しい.$k < l$は任意だったから
$$\sum_{i = 0}^{q + 1}\sum_{j< i}\langle a_0, \cdots, \hat a_{j}, \cdots, \hat a_{i}, \cdots, a_{q + 1} \rangle + \sum_{i = 0}^{q +1} \sum_{i< j}\langle a_0, \cdots, \hat a_{i}, \cdots, \hat a_{j}, \cdots, a_{q + 1} \rangle = \boldsymbol{0}$$
である.従って$\partial_{q} (\partial_{q + 1} (\sigma)) = \boldsymbol{0}$となる.$\square$

&emsp;命題1.2.3は集合の言葉で言い換えると
$$Im \partial_{q +1} \subset \ker \partial_q$$
を意味している.今までは$1$次のサイクルとバウンダリについてのみ考えていたが,上記の結果から一般の次元にも適用することが可能になる.

### 定義1.2.4 
$z \in \ker \partial_q$を$q$**次元サイクル**といい,$b \in Im \partial_{q + 1}$を$q$**次元バウンダリ**という.

&emsp;以上のことから,「中身が詰まったループのような図形」は$q$次元バウンダリとして厳密に表現できる.従って「穴になっているループのような図形」はサイクルの内バウンダリでないものとして表現できる.集合としては純粋に$\ker \partial_q - Im \partial_{q + 1}$としたいところだが,この集合は単体分割$K$の取り方に大きく変動してしまう.\
&emsp;そこで単体分割の取り方に依存せずに決まるものとして表現することを考える.まず,１つのサイクルにバウンダリを足してもまたサイクルであることは明らかである.従ってこのバウンダリの違いを無視した集合を考えたい.$Im \partial_{q +1} \subset \ker \partial_q$であったことと,この2つの集合はベクトル空間$C_q(K)$の部分ベクトル空間であることから商ベクトル空間の考えを適用できる.つまり次のように定義する.


### 定義1.2.5 
$\ker \partial_q$上に同値関係$\sim$を定義する.
$$z \sim z' (z, z' \in \ker \partial_q) \overset{\text{def}}{\Longleftrightarrow} z - z' \in Im \partial_{q + 1}$$
この同値関係による商集合を$H_q(K)$と書く.$[z]$で$z \in \ker \partial_q$の同値類を表し,$z$のホモロジー類,あるいは単に$q$次元ホモロジー類という.$H_q(K)$上には$\ker \partial_q$から次のような和が矛盾なく定義される.
$$[z] + [z'] = [z + z']$$
この和により$H_q(K)$は体$\mathbb{Z}_2$上のベクトル空間になる.この$H_q(K)$を$q$次元ホモロジー群という.

### 注意1.2.6
&emsp;一般にベクトル空間$V$とその部分空間$W$に対して,以下のような同値関係$\sim$
$$v \sim v' (v, v' \in V) \overset{\text{def}}{\Longleftrightarrow} v - v' \in W$$
を与えて得られる商集合は再びベクトル空間となり,これを商ベクトル空間と言って一般に$V/W$と書く.これに倣いホモロジー群も
$$H_q(K) = \ker \partial_q/Im \partial_{q + 1}$$
と表す.以降では$V/W$と書いたら,それは商ベクトル空間を表すものとする.

&emsp;ホモロジー類は,直感的には2つのサイクルの間の中身が詰まっているときに同一視をして得られたものであるから,事実上同値関係$\sim$はホモトピー同値を代数的に表現したものといえる.つまり次の定理が成り立つ.
### 定理1.2.6
単体複体$K,L$に対し,多面体$|K|$と$|L|$がホモトピー同値なら$H_q(K) \simeq H_q(L)$

　この定理によってホモロジー群が単体分割の違いに依らずに定まることがわかる.

&emsp; 今までは単体複体のホモロジーについて一通り述べたが,一般に単体複体の中には部分複体の構造が複数ある.ある部分複体を無視したときに得られる単体複体のホモロジーについて記述したものが相対ホモロジー群である. 相対ホモロジー群は主に後の章において使用するためここでは定義のみを述べるにとどめておくことにする.

### 定義1.2.7  相対ホモロジー群

$K$を複体として$L \subset K$を部分複体とする.このとき鎖群$C_q(K),C_q(L)$は$C_q(L) \subset C_q(K)$を満たす.そこで$q$**次元相対鎖群**$C_q(K, L)$を 
$$C_q(K, L) := C_q(K) / C_q(L)$$
とおく.
$\partial_q: C_q(K) \longrightarrow C_{q - 1}(K)$を境界準同型とすると$\partial_q(C_q(L)) \subset C_{q - 1}(L)$であるから
$$c - c' \in C_q(L) (c , c' \in C_q(K)) \Longrightarrow \partial_q(c) - \partial(c') \in C_{q - 1}(L)$$
が成り立つ.従って相対鎖群の間の境界準同型$\partial_q^*:C_q(K, L) \longrightarrow C_{q - 1}(K, L)$が
$$\partial_q^*([c]) := [\partial_q(c)]$$
として矛盾なく定義される.\
$\partial_{q + 1} \circ \partial_q = \boldsymbol{0}$より$\partial_{q + 1}^* \circ \partial_q^*= \boldsymbol{0}$であるから,鎖群のときと同様に
$$Im \partial_{q + 1}^*\subset  \ker  \partial_q^* $$
が成り立つ.
$$H_q(K, L) : = \ker  \partial_q^*/Im \partial_{q + 1}^*$$
としたものを **$q$次元相対ホモロジー群** という.

### 定義1.1.5 ホモロジー多様体
位相空間$M$が単体分割可能であり,その単体分割を$t:K \longrightarrow M$とし,$p \in M$とする.
$$H_q(L_M(p)) := H_q(L_K(t^{-1}(p)))$$
とおくとき,これを$M$の点$p$における**局所ホモロジー群**という.\
さらに,ある自然数$n$が存在して,任意の点$p \in M$に対して
$$H_q(L_M(p)) \simeq H_q(S^{n -1}) \ (q = 0, 1, \cdots)$$
が常に成り立つとき,$M$を **$n$次元ホモロジー多様体** という.

### 例1.1.6
単体分割可能な可微分多様体はホモロジー多様体である.

### 定理1.1.7
$n$次元ホモロジー多様体$M$の単体分割を$t:K \longrightarrow M$とすると,一般に次の性質が成り立つことが知られている.
$$(i) \dim K = n$$ 
$$(ii) \forall \sigma \in K \exists \tau \in K s.t. \sigma \prec \tau , \dim \tau = n$$
$$(iii) \sigma \in K \dim \sigma = n - 1 \Longrightarrow \# \{\tau \in K | \sigma \prec \tau, \dim \tau = n \} = 2$$

## 1.2向きづけ可能性

### 定義1.2.1
向きのついた$q$単体$\langle a_0,\cdots, a_q \rangle$に対し,向きのついた$q - 1$辺単体$(-1)^i \langle a_0, \cdots, \hat a_i, \cdots, a_q \rangle$の向きを$\langle a_0,\cdots, a_q \rangle$から**導かれる向き**という.

### 定義1.2.2
二つの$q$単体$\sigma, \tau$の共通部分$\gamma = \sigma \cap \tau $が$\sigma$と$\tau$とに共通の$q-1$辺単体であり,$\sigma$と$\tau$にそれぞれ向き$\langle \sigma \rangle$と$ \langle \tau \rangle$を決めたとする.\
$\langle \sigma \rangle $から導かれる$\gamma$の向きと,$ \langle \tau \rangle$から導かれる$\gamma$の向きが逆になっているとき,二つの向き$\langle \sigma \rangle$と$ \langle \tau \rangle$は**同調**しているという.

### 定義1.2.3
$n$次元ホモロジー多様体$M$の単体分割を$t:K \longrightarrow M$とすると,上述の定理1.1.2の性質$(iii)$から$K$の$n-1$単体はちょうど二つの$n$単体の辺単体となっている.$K$の全ての$n$単体に向きを定め,共通の$n-1$単体を持つ任意の二つの$n$単体の向きが同調しているようにできるとき,$n$次元ホモロジー多様体$M$は**向きづけ可能**であるという.

### 定理1.2.4
弧状連結な$n$次元ホモロジー多様体$M$において次が成り立つ.\
$(i)H_n(M) = 0 \ or \  \mathbb{Z}_2$\
$(ii)M$が向きづけ可能である$\Longleftrightarrow H_n(M)\simeq \mathbb{Z}_2$

弧状連結で向づけ可能な$n$次元ホモロジー多様体$M$に対して,定理1.2.4$(ii)$から得られる$H_n(M)$の生成元を一つ定めることを$M$に**向きをつける**と言い,この生成元を$M$の**基本ホモロジー類**と言う.


# 2.コホモロジー
&emsp;ホモロジーの双対の概念としてコホモロジーがある.コホモロジーは多様体の向きづけ可能性を議論する際によく使用される概念である.大雑把に言うと単体複体内の単体を線形(準同型)写像として捉え直したことで得られるベクトル空間である.
&emsp;本論文の目的であるホイットニー類のパーシステント化をする際にもやはりこのコホモロジーを使用する.しかしコホモロジーはホモロジーに比べてやや抽象度が高い.そこでこの章ではコホモロジーの定義について述べた後,それが前述したホモロジーとどのような関係性を持つのかについても記述する.より詳しくは普遍係数定理やポアンカレ双対定理について述べることにする.

## 2.1コホモロジー群


### 定義2.1.1 コホモロジー群
単体複体$K$の$q$次元鎖群$C_q(K)$から$\mathbb{Z}_2$への準同型写像全体を$C^q(K)$とする.$C^q(K)$上に加法$f + g$とスカラー倍$mf$を以下で定め,ベクトル空間にしたものを$q$**次元双対鎖群** という:
$$(f + g)(c) := f(c) + g(c) \\ (mf)(c) := mf(c)$$
ただしここで$f,g \in C^q(K),m \in \mathbb{Z}_2, c \in C_q(K)$としている.さらに$\partial_{q + 1}:C_{q + 1}(K) \longrightarrow C_{q}(K)$を境界準同型とするとき,**$q$次元双対境界準同型** $\delta_q:C^q(K) \longrightarrow C^{q + 1}(K)$を
$$\delta_q(c) = f \circ \partial_{q + 1} \ \ (f \in C^q(K))$$ 
によって定義する.
$\partial_{q} \circ \partial_{q + 1} = 0$だから
$$\delta_{q} \circ \delta_{q - 1} = 0$$
が成り立つ.つまり$Z^q(K) = Ker \delta_q \ \ , B^q(K) = Im \delta_{q - 1}$とすると
$$B^q(K) \subset Z^q(K)$$
が成り立つ.\
そこで$q$**次元コホモロジー群**$H^q(K)$を以下で定義する.
$$H^q(K) := Z^q(K)/B^q(K)$$

&emsp;1では相対ホモロジーを導入したが,同じようにコホモロジーに対しても相対コホモロジーを考えることができる.これはいわば相対ホモロジーの双対である.
### 定義2.1.2 相対コホモロジー群

相対鎖群$C_q(K, L)$から$\mathbb{Z_2}$への準同型写像全体の集合を$C^q(K, L)$と書く.$\partial_q^*:C_q(K, L) \longrightarrow C_{q - 1}(K, L)$を相対鎖群の境界準同型とするとき,その双対境界準同型$\delta_q^*:C^q(K, L) \longrightarrow C^{q + 1}(K, L)$を
$$\delta_q^*(f) : = f \circ \partial_{q + 1}^* \ \ \ (f \in C^q(K, L))$$
で定義する.$\partial_q^*\circ \partial_{q + 1}^* = \boldsymbol{0}$であるから,$\delta_{q} ^* \circ \delta_{q - 1}^*= \boldsymbol{0}$である.つまり
$$Im \ \delta_{q - 1}^* \subset \ker  \delta_{q}^*$$
が成り立つ.そこで相対コホモロジー群$H^q(K, L)$を
$$H^q(K, L) := \ker  \delta_{q}^*/ Im \delta_{q - 1}^*$$
で定義する.

### 定義2.1.2 双対基底
$C_{n}(K)$の基底を$\sigma_1, \cdots, \sigma_m$として,その双対を$\hat \sigma_1,\cdots,\hat \sigma_m \in C^n(K )$とする.すなわち
$$\hat \sigma_i (\sigma_j) = \delta_{ij}  \ \ \ (0 \leq i, j \leq m)$$
となる準同型とする.このとき$\hat \sigma_1, \cdots, \hat \sigma_m $は$C^n(K )$の基底をなす.これを$C_{n}(K )$の基底$\sigma_1, \cdots, \sigma_m$に対する**双対基底**という.
$c = c_1 \sigma_1 + \cdots + c_m \sigma_m \in C_n(K ), \ (c_i \in \mathbb{Z}, i = 1, \cdots, m)$とするとき
$$\hat \sigma_i (c) = \sum_{j = 1}^{m}c_j \hat \sigma_i (\sigma_j) = \sum_{j = 1}^{m}c_j \delta_{ij} = c_i$$
となるから$c \in C_n(K )$は双対基底における値$\hat \sigma_i (c) \ (i = 1, \cdots, m)$によって一意に決定される.$\hat \sigma_1, \cdots, \hat \sigma_m $は$C^n(K )$の基底であったから,各$f \in C^n(K )$に対して値$c_f \in \mathbb{Z}$を割り当てることによって,$f(c) = c_f$となるような$c \in C_{n}(K )$を一意的に構成できる.



### 定理2.1.3 ポアンカレ双対定理
弧状連結で向づけ可能な$n$次元ホモロジー多様体$M$に対し
$$H^q(M) \simeq H_{n - q}(M) \ \ (q = 0, 1, \cdots)$$
であり,その同型写像$l_q:H^q(M) \simeq H_{n - q}(M)$は定義2.2.4のキャップ積と$M$の基本ホモロジー類$[M]$を用いて
$$l_q([z]) = [M] \frown [f]$$
で与えられる.



## 2.2積

### 定義2.2.2 カップ積
$p$次元双対鎖$c \in C^p(K )$と$q$次元双対鎖$d \in C^q(K )$に対して$c \smile d \in C^{p + q}(K )$を得る操作を**カップ積**と言い,以下で定義する.
$$(c \smile d)(\langle a_0, \cdots, a_{p + q} \rangle) := c(\langle a_0, \cdots , a_p \rangle) \cdot d(\langle a_{p}, \cdots , a_{p+q} \rangle )$$
このとき
$$\delta_{p + q}(c \smile d) = (\delta_{p} c) \smile d + c \smile \delta_q d$$
を満たすからこれによってコホモロジー類同士のカップ積$H^p(K ) \times H^q(K ) \longrightarrow H^{p + q}(K )$が矛盾なく定義される.
$$\lbrack z \rbrack \smile \lbrack z' \rbrack =\lbrack z \smile  z' \rbrack$$


### 定義2.2.4 キャップ積
$p + q$次元鎖$c \in C_{p+q}(K )$と$f \in C^p(K )$に対して,そのキャップ積$c \frown f \in C_q(K )$を以下で定義する.
$$ g(c \frown f) := (f \smile g)(c)  \ \ , g \in C^q(K )$$
キャップ積は以下の性質を持つ.
$$\partial (c \frown f) = \{(\partial c \frown f) - c \frown \delta f\} $$ 
これによってホモロジー類とコホモロジー類のカップ積$H_{p + q}(K ) \times H^p(K ) \longrightarrow H_{q}(K )$が矛盾なく定義される.
$$\lbrack z \rbrack \frown \lbrack f \rbrack =\lbrack z \frown f \rbrack$$

加群の係数の環として$\mathbb{Z}$を用いたが,$\mathbb{Z}/2\mathbb{Z}$としても同様の議論が可能である.以下では$\mathbb{Z}/2\mathbb{Z}$を係数とする場合は$C_*(K \mathbb{Z}/2)$などと表す.ホモロジー群や双対鎖群,コホモロジー群についても同様である.


## 2.3 ホイットニー類

### 定義2.3.1 ベクトル束
$E, B$を位相空間として,$\pi: E \longrightarrow B$を連続な全射とする.今,任意の$b_0 \in B$に対して次の$(i),(ii)$を満たすとき,$\xi = (B, E,\pi)$をベクトル束という.\
$(i) F_{b_0}(\xi ) := \pi^{-1} (b_0)$はベクトル空間である.\
$(ii) b_0$における$B$のある開近傍$U \subset B$と同相写像$h:U \times \mathbb{R}^n \longrightarrow \pi^{-1}(U)$が存在して,対応
$$\mathbb{R}^n \ni x \longmapsto h(b, x) \in F_b(\xi)$$
が各$b \in U$においてベクトル空間としての同型を与える.
$E$をベクトル束の全空間,$B$を底空間,$\pi$を射影といい$E(\xi), B(\xi), \pi_\xi$などと書く.


### 例2.3.2
多様体$M$の接束$TM$はベクトル束である.
### 例2.3.3
実射影空間$P^n = \{{x, -x}| x \in S^n \}$を底空間とする**直線束**と呼ばれるベクトル束$\gamma _n^1 = (P^n, E, \pi)$は次のように定義される.
$$F_{\{x, -x\}}(\gamma_n^1)= \{(\{x, -x\}, \lambda x)|  \lambda \in \mathbb{R}\}$$
$$E = \cup_{b \in P^n}F_b(\gamma_n^1)$$
$$\pi(F_b(\gamma_n^1)) = b$$


### 定義2.3.4 束写像
全空間の間の連続写像$f:E(\xi) \longrightarrow E(\eta)$が,各$F_b(\xi)$に対してある$F_{b'}(\eta)$と線型同型を与えるとき,$f$を束写像という.$\hat f(b) = b'$とすると$f$は連続写像$\hat f: B(\xi) \longrightarrow B(\eta)$を与える.

# 3.パーシステント性
あるパラメータ例えば時間によって単体複体が増大していく際,その多面体内の穴が,ある時刻で生まれてさらに進んだ時刻で消えるといった現象が観察される.パーシステントホモロジー(あるいはそのパーシステント図)はそういった現象をホモロジー群の生成元によって記録するものである.\
しかし,通常のパーシステントホモロジーは向き付け可能性の可否を判別することはできていない.本論文ではこの問題を解消するために,パーシステント性をコホモロジーのレベルに発展させたパーシステントコホモロジーおよびパーシステントホイットニー類を構成する.

### パーシステントホモロジー

### パーシステント図

### 定義3.1.1 パーシステントコホモロジー
単体複体の増大列
$$\mathbb{K}:K_0 \subset K_1 \subset \cdots \subset K_T$$
を考える.包含写像$\phi_{j}^{j + 1}: K_j \longrightarrow K_{j + 1}$によってホモロジー群へ誘導される準同型
$$(\phi_{j}^{j + 1})^*: H^p(K^{j + 1}) \longrightarrow H^p(K^{j})$$
が得られる.$j = 0, \cdots , T$まで考えることで得られる系列
$$H^p(K_0) \longleftarrow H^p(K_1) \longleftarrow \cdots \longleftarrow H^p(K_T)$$
を**パーシステントコホモロジー**という.また,
$$PH^q(\mathbb{K}) := H^q(K_T) \oplus \cdots \oplus H^q(K_0)$$
と定義した群をパーシステントコホモロジー群と言う.ここで直和の順番はフィルトレーションのパラメータに関して降順にしている.

パーシステントコホモロジー群には次の定義3.1.2に定める$\mathbb{Z}_2[x]$との演算を考える.

### 定義3.1.2 
$r = (r_T, \dots, r_0) \in PH^q(\mathbb{K}) \ \ (r_i \in H^q(K_i))$とダミー変数$x$との演算を次のように定義する.
$$x \cdot r := (0, (\phi_{T - 1}^{T})^*(r_T), \dots , (\phi_{0}^{1})^*(r_1))$$
ただし,ここで$\phi_{i}^{i + 1}: K_i \hookrightarrow K_{i + 1}$は包含写像としている.$(\phi_{i}^{i + 1})^*(r_{i + 1}) \in H^q(K_{i + 1})$だから$x \cdot r \in PH^q(\mathbb{K})$である.
このとき$r' \in PH^q(\mathbb{K})$とすると
$$x \cdot (r + r') = x \cdot r + x \cdot r'$$
が成り立つ.ここでさらに$f, g \in \mathbb{Z}_2[x]$に対して再帰を用いて
$$1 \cdot r = r$$
$$(f g) \cdot r := f \cdot (g \cdot r)$$
$$(f + g) \cdot r := f \cdot r + g \cdot r$$
と定義すると,$PH^q(\mathbb{K})$は$\mathbb{Z}_2[x]$上の加群となる.

以降は$x \cdot r$を$xr$と書くことにする.

### 定義3.1.2
同じ長さの２つのフィルトレーション
$$\mathbb{X}: \emptyset = X_0 \subset X_1 \subset \cdots \subset X_m = X$$
$$\mathbb{Y}: \emptyset = Y_0 \subset Y_1 \subset \cdots \subset Y_m = Y$$
に対して,それぞれの包含写像を$\phi_i^{i + 1}: X_i \hookrightarrow X_{i + 1}, \psi_i^{i + 1}: Y_i \hookrightarrow Y_{ i + 1}$とする.今連続写像の族$f_i: X_i \longrightarrow Y_i$が存在して,
$$\psi_i^{i + 1} \circ f_i = f_{i + 1} \circ \phi_{i}^{i + 1} \ \ ( i = 1, \cdots , m - 1)$$
が成り立つとき,$f = \{f_i\}$を**フィルトレーションの間の連続写像**といい,$f: \mathbb{X} \longrightarrow \mathbb{Y}$と書くことにする.

### 命題3.1.3
同じ長さの２つのフィルトレーション
$$\mathbb{X}: \emptyset = X_0 \subset X_1 \subset \cdots \subset X_m = X$$
$$\mathbb{Y}: \emptyset = Y_0 \subset Y_1 \subset \cdots \subset Y_m = Y$$
から定まるパーシステントホモロジー群をそれぞれ$PH^q(\mathbb{X}) = \oplus H^q(X_i), PH^q(\mathbb{Y}) = \oplus H^q(Y_i)$とし,$\{f_i\}: \mathbb{X} \longrightarrow \mathbb{Y}$をフィルトレーションの間の連続写像とする.各$f_i$が引き起こすコホモロジーの間の準同型$f_i^*:H^q(Y_i) \longrightarrow H^q(X_i)$は各$r_{i + 1} \in H^q(Y_{i + 1})$に対して以下の性質を満たす.
$$xf_{i + 1}^*(r_{i + 1}) = f_{i}^*(x r_{i + 1})$$

### 定義4.1.1
$E$を全空間,$X$を底空間とするベクトル束$\xi = (E,X, \pi)$を考える.今$X$に関するフィルトレーション
$$\mathbb{X}:\emptyset = X_0 \subset X_1 \subset \cdots \subset X_m = X$$
を考えると,各$i\geq 1$に対して$X_i$を底空間に対する$\xi$の制限束$\xi|X_i = (E_i, X_i, \pi_i)$が定義される.また,さらに
$$E_1 \subset \cdots \subset E_m = E$$
が成り立つ.ベクトル束の系列$\xi|X_i(i = 1, \cdots, m)$を底空間のフィルトレーションから誘導されるベクトル束のフィルトレーションといい,記号$\xi|\mathbb{X}$と書くことにする.
