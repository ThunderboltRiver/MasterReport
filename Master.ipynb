# ホモロジー
位相幾何学では,球体やトーラスなどのユークリッド空間内の図形から連結成分や輪っかや空洞の数と言った幾何学的な特徴を捉える際に,一度図形を簡単な多面体に近似(変形)して,代わりにこの多面体の特徴を捉えるといった手法を用いる.しかし雑に近似したところで,元の図形の関心のある特徴が大きく崩れては意味がない.そこで用いられる主な近似の手法として,位相同型やホモトピー同値が挙げられる.これら二つの変形によって,図形の見た目を大きく変えたとしても関心のある特徴を保存することができる.例えば,連結成分や輪っかや空洞の数について関心があれば元の図形とホモトピー同値な多面体を調べれば良い.それよりももっと多くのことを調べる際は,ホモトピー同値よりも近似度合いの強い位相同型な多面体について調べれば良い.このようにして考察の対象を多面体のみに狭めることができる.\
そこでこのセクションでは多面体とは何かについて述べた後,連結成分や輪っかや空洞の数を調べる際の代数学的な概念であるホモロジー群について解説する.その後,本論文での考察対象となるホモロジー多様体について解説する.

## 単体複体

### 単体複体と多面体
多面体は,単体と呼ばれる簡単な図形がいくつか合併したものと捉え,その単体の集まりが単体複体として定義される.そこで先に単体と単体複体について解説する.

#### 定義
$R^N$内の$n + 1$個の点$a_0, \cdots , a_n$に対して,ベクトル$\vec{a_0a_1}, \cdots, \vec{a_0a_n}$が一次独立である(一般の位置にある)とき,集合
$$|a_0, \cdots, a_n| = \{\sum_{i = 0}^{n} \lambda_i a_i | \lambda_i \geq 0 (i = 0, \cdots, n) \sum_{i = 0}^{n}\lambda_i = 1\}$$
を,$a_0, \cdots, a_n$を頂点とする **$n$次元単体** といい,$\sigma = |a_0, \cdots, a_n|$などと書く.\
$\sigma$が単体であるとき,$\sigma$の頂点集合$\{a_0, \cdots, a_n\}$の空でない部分集合を頂点とする単体$\tau$を$\sigma$の**辺単体**といい,$\tau \preceq \sigma$と書く.単体$\sigma$が$n$次元であることを$\dim \sigma = n$と書く.

#### 定義
$R^N$内のいくつかの単体からなる有限集合$K$であって,以下の条件を満たすものを **単体複体** という.
$$(i) \sigma \in K, \tau \preceq \sigma \Longrightarrow \tau \in K$$
$$(ii) \sigma, \tau \in K \sigma \cap \tau \ne \emptyset \Longrightarrow  \sigma \cap \tau \preceq \sigma, \tau$$
ここで空集合$\emptyset$も単体複体であると定義する.$K \ne \emptyset$のときは$K$の$0$単体は **$K$の頂点** とよばれ,それら全体の集合を$\hat K$などと書く. $n:= \max{\{\dim \sigma | \sigma \in K \}}$を単体複体 $K$ の**次元**といい,$\dim K = n$と書く.また,単体複体$K$の全ての単体の和集合
$$|K| :=  \bigcup_{\sigma \in K} \sigma$$
を$K$の**多面体**という.$K = \emptyset$のときは$K$の次元は定義せず,多面体は$\emptyset$と定義する.

ホモロジー多様体や後のセクションで述べるパーシステント性において必要になる部分複体とその具体例について述べる.

#### 定義
単体複体$K$の部分集合$L \subset K$がまた単体複体であるとき,$L$を$K$の **部分複体** という.\
$L \subset K$が部分複体であるための必要十分条件は,定義1.1.2の条件$(i)$を満たすことである.

#### 例
$K$を単体複体とし,$p \in |K|$を多面体の点とする.
$$S_K(p) := \{\tau \in K | \exists \sigma \in K s.t. \ p \in \sigma , \tau \prec \sigma \}$$
$$L_K(p):= \{ \tau \in S_K(p)| p \notin \tau \}$$
とおくと,これらはともに$K$の部分複体である.それぞれ **$K$における$p$の星状複体,まつわり複体** という.

&emsp;直感的な言い方をすると,$p \in |K|$のまつわり複体とは,多面体$|K|$内で$p$に触れることなく囲うように位置する$K$の単体を全て集めた単体複体といえる.

### 単体複体の組み合わせ構造
&emsp;多面体を構成するには単体複体を構成する必要があるが,単体はユークリッド空間の図形であるからいちいちこれを作り上げるのは手間がかかる.また,多面体の辺の長さや面積,体積,内角といった量に元から関心がないなら,一般の位置にある点を具体的に考える必要はない.\
&emsp;例えば多面体として三角形を何か一つ構成する際,どんな形だろうととにかく三角形でありさえすれば良い場合は具体的な頂点を取る必要がなく,あくまで$3$つの頂点が一般的な位置にあるという情報だけで事足りる.\
&emsp;この視点で単体複体の頂点全体に注目すると,各単体はその頂点の組み合わせ(部分集合)として捉えられることがわかる.より具体的には次の命題1.1.5が成り立つ.

#### 命題
$K$を単体複体とする.$K$の頂点全体の集合$\hat K$の部分集合族$\Sigma$を,$K$の単体の頂点集合全体として定義する.
$$\Sigma := \{\{a_0, \cdots, a_k \} \subset \hat K | k \geq 0, |a_0, \cdots, a_k| \in K \} \cup \{ \emptyset \}$$
このとき$\hat K$と$\Sigma$は次のような性質を持つ.
$$(i) \forall v \in \hat K, \{v\} \in \Sigma $$
$$(ii)\hat \sigma \in \Sigma, \hat \tau \subset \hat \sigma \Longrightarrow \hat \tau \in \Sigma $$
$proof)$
$(i):v \in \hat K$は$K$の頂点,つまり$0$単体であるから$|v| = v \in K$である.従って$\{ v \} \in \Sigma$\
$(ii):\hat \sigma \in \Sigma$を頂点集合とする$K$の単体を$|\hat \sigma|$とする.$\hat \tau \subset \hat \sigma$が空でないなら,$|\hat \tau| \prec |\hat \sigma|$である.$K$が複体であることから,定義1.1.2の条件$(i)$より$|\hat \tau| \in K$よって$\hat \tau \in \Sigma$ .$\square$

&emsp;命題1.1.5は幾何学的な概念である単体複体から条件$(i), (ii)$にあるような組み合わせ論的な構造を抽出できることを示している.そこで次のように定義する.
#### 定義
有限集合$V$とその部分集合族$\mathcal{S}$が以下の条件を満たすとき,対$(V, \mathcal{S})$を**抽象複体**と言う.
$$(i) \forall v \in V, \{ v \} \in \mathcal{S}$$
$$(ii) \hat \sigma \in \mathcal{S}, \hat \tau \subset \hat \sigma \Longrightarrow \hat \tau \in \mathcal{S} $$
このとき$V$は **頂点集合** と呼ばれ,その元を **頂点** という.$\mathcal{S}$ の元$\hat \sigma$は,その元の個数が$n + 1$であるなら $n$**次元単体** と呼ばれ,$\dim \hat \sigma = n$と書く.


#### 例
距離空間$(X, d)$の有限個の点からなる集合$P = \{x_1, \cdots, x_n\}$と$r >0$に対して,抽象複体$VR_r(P) = (V, \mathcal{S})$を以下で定義する.
$$ V := \{1, \cdots, n \} \\ \mathcal{S} := \{ B \subset V |  \forall i,j \in B, d(x_i, x_j) \leq 2r \}$$
$VR_r(P)$を**ヴィートリスリップス複体**という.

&emsp;抽象複体という視点に立つと$n$単体は$n + 1$個の元からなる集合として捉えられ,それらの集合族によって単体複体の大まかな形が実現されていると捉えることができる.  
&emsp;次に二つの抽象複体の間に定義される写像であって,単体を単体へ対応させるものについて定義する.

#### 定義
二つの抽象複体$(V, \mathcal{S}), (V', \mathcal{S}' )$の頂点集合の間の写像$\phi: V \longrightarrow V' $は,以下の条件を満たすとき単体写像という.
$$\hat \sigma \in \mathcal{S} \Longrightarrow \phi(\hat \sigma) \in \mathcal{S}'$$
このとき$\phi: (V, \mathcal{S}) \longrightarrow (V', \mathcal{S}')$などと書く.また,二つの単体複体$K,L$に命題1.1.5を適用することで得られる抽象複体の間に単体写像を構成したときは単に$\phi:K \longrightarrow L$などと書く.\
単体写像は単射であるとは限らないから,写った先の単体の次元は,元の単体より少ない可能性がある.もし単射であるなら次元は保たれる.\
この単体写像を用いて抽象複体の同型について定義する.これは幾何学的は,二つの単体複体に同じ組み合わせ的構造があることを示すものである.

#### 定義
単体写像$\phi: (V, \mathcal{S}) \longrightarrow (V', \mathcal{S}')$は,全単射であり,その逆写像もまた単体写像であるとき,**単体同型写像** という.
さらにこのとき,二つの抽象複体$(V, \mathcal{S}),(V', \mathcal{S}')$は**同型**であるという.

命題1.1.5は幾何学的な単体複体から抽象複体を構成したが,この逆もまた成り立つ.
#### 命題
抽象複体$(V, \mathcal{S})$に対して,命題1.1.5の$(\hat K, \Sigma)$と同型になるような単体複体$K$を構成できる.
この$K$を抽象複体$(V, \mathcal{S})$の**幾何学的実現**という.\
$proof)V = \{v_0, \cdots, v_N \}$とするとき,$\mathbb{R}^N$の標準基底$\boldsymbol{e}_i (i = 1, \cdots, N)$を取る.$\boldsymbol{e}_0 = \boldsymbol{0}$とするとき,$\boldsymbol{e}_0, \cdots, \boldsymbol{e}_N$は一般の位置にある.そこで単体複体$K$を次で定義する.
$$K := \{|\boldsymbol{e}_{i_0}, \cdots, \boldsymbol{e}_{i_k}| \ | \{v_{i_0}, \cdots, v_{i_k} \} \in \mathcal{S},  \{i_0, \cdots, i_k\} \subset \{0, \cdots, N\}, k \geq 0 \} \}$$
この$K$に命題1.1.5を適用して得られた$(\hat K, \Sigma)$と抽象複体$(V, \mathcal{S})$との間に写像$\phi:V \longrightarrow \hat K$を$\phi(v_i) = \boldsymbol{e}_i$で定義するとこれが同型な単体写像となる.$\square$

&emsp; 命題1.1.5と命題1.1.9により,単体複体を構成する際は抽象複体のみを考えるだけで十分であることがわかる.

### 単体分割
&emsp;以上で多面体の解説を終えるが,最後に多面体を一般の図形に近似するための単体分割について述べる.

#### 定義
2つの位相空間$X, Y$が位相同型であるとは,2つの連続写像$f:X \longrightarrow Y, g:Y \longrightarrow X$が存在して
$$g \circ f = id_{X} \ \ \ f \circ g = id_{Y}$$
となることをいう.ここで$id_{X}$は位相空間$X$の恒等写像である.

#### 定義
位相空間$M$が,ある単体複体$K$の多面体と位相同型であるとき,同相写像$t:K \longrightarrow M$を**単体分割**といい,$M$は単体分割可能であるという.

単体分割によって一般の図形を単体複体に置き換えて考察することが可能になる.

### 脈体定理

#### 定義
位相空間$X$がその部分集合族$\Phi = \{F_i | i = 1 , \cdots, m\}$によって被覆
$$X = \bigcup_{i = 1}^m F_i$$
されているとする,このとき以下で定義される抽象単体複体$(V, \mathcal{S})$を$\Phi$の**脈体**といい$\mathcal{N}(\Phi)$で表す.
$$V:= \{1, \cdots, m\} \\ \mathcal{S} = \{ S \subset V | \bigcap_{i \in S} F_{i} \ne \emptyset\}$$

#### 例
距離空間$(X, d)$の有限個の点からなる集合を$P={x_1, \cdots, x_n}$とする.半径$r > 0$の閉球体の集まり$\Phi = \{B(x_i, r) | i = 1, \cdots, n\}$の脈体$\mathcal{N}(\Phi)$を半径$r$の**チェック複体**といい$C_r(P)$で表す.

#### 例
距離空間$(X, d)$の有限個の点からなる集合を$P = \{x_1, \cdots, x_n \}$する.各$x_i$のボロノイ領域$V(x_i) := \{ x \in X | d(x, x_i) \leq d(x, x_j)  \ \forall j \}$と半径$r$の閉球体$B(x_i, r)$の共通部分を$(x_i, r) := V(x_i) \cap B(x_i, r)$とするとき,$\Phi = \{W(x_i, r) | i = 1, \cdots, n\}$の脈体を半径$r$の**アルファ複体**といい,$\alpha_r(P)$で表す.

#### 定理
(脈体定理)
$X$を位相空間とし,$\Phi = \{F_i | i= 1, \cdots, m \}$をその開(閉)被覆
$$X = \bigcup_{i =1}^m F_i$$
とする.$\Phi$のいくつかの集合による空でない交差が一点に可縮であるならば$\Phi$の脈体$\mathcal{N}(\Phi)$の幾何学的実現は$X$にホモトピー同値である.つまり
$$X \simeq | \mathcal{N}(\Phi)|$$
が成り立つ.

## ホモロジー群

### 複体のホモロジー
&emsp;多面体内に$n$次元の穴がいくつあるかと言ったことを代数学的に調べる手法としてホモロジー群がある.ここでは一旦,ホモロジー群の直感的な説明を簡単に行う.\
&emsp;多面体に含まれる図形は,単体複体内のいくつかの単体の和集合として大雑把に表現できる.従って,多面体内の穴もまたいくつかの単体の和集合として解釈できる.しかし,穴の情報だけが欲しいのなら,ループのような形をした図形のみを考えればよく,さらに言えば,「中身(内部)が埋まっているなら無視する」という規則を設けた方が便利である.\
&emsp;上記の考え方は代数学的に厳密に表現することが可能で,特に「中身(内部)が埋まっているなら無視する」という規則は商群によって表現することができる.この商群をホモロジー群という.\
このセクションでは,これをもう少し厳密な言い方で表現する.そのためにまずは集合論的な和集合を,代数学的な形式和として捉え直すことから始める.これが鎖群の考え方である.

#### 定義
$K$を空でない単体複体とする.$K$の$q$次元単体全体を基底とする,体$\mathbb{Z}_2$上のベクトル空間を$K$の**$q$次元鎖群**といい$C_q(K)$と書く.つまり,$\sigma_1^{q}, \cdots, \sigma_{k_q}^q$を$K$の$q$次元単体全体とするとき
$$C_q(K) := \{\sum_{i = 1}^{k_q}m_i \sigma_i^{q}| m_i \in \mathbb{Z}_2 \ , i = 1, \cdots, k_q\}$$
と定義する.$q > \dim K$と$q < 0$のときは$C_q(K) = \boldsymbol{0}$と定義する.また$K = \emptyset$のときは任意の$q$に対して$C_q(K) = \boldsymbol{0}$と定義する.ここで$c \in C_q(K)$を**$q$次元鎖**という.\
&emsp;鎖群の構成によって,集合論的な単体の和集合は鎖として表現される.ただし,鎖として単体を表すときは$\sigma = \langle a_0, \cdots, a_q \rangle$などと書くことにする.


次に「ループのような図形」や「中身が詰まっている」という表現をより詳細化するために境界準同型という鎖群間の写像を定義する.
#### 定義
整数$q$を$1 \leq q \leq \dim K$とするとき,$K$の$q$次元単体$\sigma =  \langle a_0, \cdots, a_q \rangle \in C_q(K)$の境界$\partial_q (\sigma) \in C_{q - 1}(K)$を以下で定義する.
$$\partial_q (\sigma) := \sum_{i = 0}^{q}\langle a_0, \cdots,a_{i - 1}, a_{i + 1}, \cdots a_q \rangle$$
さらに一般の$q$次元鎖$c = \sum_{i = 1}^{k_q} m_{i} \sigma_i^q \in C_q(K)$に対しては
$$\partial_q(c) := \sum_{i = 1}^{k_q}m_i \partial_q (\sigma_i^{q})$$
と定義する.ただし,ここで$\sigma_i^{q} (i =1, \cdots, k_q )$は単体複体$K$の$q$次元単体全体とする.\
$q > \dim K$あるいは$q \leq 0$となる$q$に対しては
$$\partial_q(c) := \boldsymbol{0} \ \ c \in C_q(K)$$
と定義する.
これによって任意の$q \in \mathbb{Z}$に対して準同型写像
$$\partial_q : C_q(K) \longrightarrow C_{q - 1}(K)$$
が定義される.これを **$q$次元境界準同型写像** という.


　以上の準備のもとで「ループのような図形」をより厳密に「どの点からも始点と終点が同じであるように一筆書きできる図形」として定義し,その性質を考察することにする.\
　議論を簡単にするために一旦具体的に円周$S^1$についてのみ考える.まず$S^1$上に異なる３点をとることで単体分割が得られるが,そのどの頂点もちょうど2つの辺の頂点となっていることがわかる.これは点の数をどれだけ増やしていっても観測される.一方で,別の$S^1$と何処かの一点で同一視してできた図形(8の字ような図形)はどうなるだろうか.この場合,同一視された１点と他の異なる2点を,2つの$S^1$上からそれぞれ取ることで単体分割が得られ,そのどの頂点も偶数本の辺の頂点となっていることがわかる.\
　以上の考察から,位相空間$M$の内のループのような図形は,単体分割$t:K \longrightarrow M$とその鎖群$C_q(K)$の表現を用いると,次の条件を満たす鎖$c \in C_1(K)$として一般化できる.
$$\partial_1(c) = \boldsymbol{0}$$

そこで上記の性質を持つ鎖を**単体複体$K$の$1$次のサイクル**と呼ぶことにする.こうして「ループのような図形」は1次のサイクルとして厳密に表現できた.\
　しかし,一口に$1$次のサイクルといっても「中身が詰まっている」ものと「穴になっている」ものがある.たとえば円板$D^2$は$S^1$を含むから$1$次のサイクルを持っているが,同時に「中身が詰まっている」図形でもある.一方で$S^1$自身は「穴になっている」図形である.よって次にこれを区別する方法を考える.\
　簡単のため$D^2$に対して考察する.$D^2$の境界上に異なる点を３つ取ると,$D^2$はあたかも$2$単体を伸ばしてできた図形として見做せる.つまり$D^2$の単体分割として,$2$単体の辺単体全体が考えられる.この単体分割において$2$単体に境界準同型$\partial_2$を施すことで$D^2$の部分集合としての$S^1$を表す鎖が得られる.\
　以上の考察から,中身が詰まっている鎖$c \in C_1(K)$とは,
$$\exists c' \in C_2(K) \ s.t. \ \partial_2(c') = c$$
となるものであると解釈できる.このような鎖を**単体複体$K$の$1$次のバウンダリ**と呼ぶことにする.上記の例では,$D^2$の部分集合としての$S^1$を表す鎖が$1$次のバウンダリであるが,特に$1$次のサイクルでもある.これは偶然のように思われるかもしれないが,一般に成り立つことである.このことを次の命題に示す.

#### 命題
境界準同型$\partial_q:C_q(K) \longrightarrow C_{q - 1}(K)$と任意の$q$に対して
$$\partial_{q} \circ \partial_{q + 1} = \boldsymbol{0}$$
が成り立つ.\
$proof)q > \dim K$と$q \leq 0$のときは明らかだから$0 \leq q \leq \dim K$のときを示す.$\sigma = \langle a_0, \cdots, a_{q+1} \rangle$を$K$の単体とする.準同型性から$\partial_{q} (\partial_{q + 1} (\sigma)) = \boldsymbol{0}$を示せば良い.
$$\partial_q(\partial_{q +1}(\sigma)) = \partial_q(\sum_{i = 0}^{q + 1}\langle a_0, \cdots, \hat a_{i}, \cdots, a_{q + 1} \rangle) \\ = \sum_{i = 0}^{q + 1}\sum_{j< i}\langle a_0, \cdots, \hat a_{j}, \cdots, \hat a_{i}, \cdots, a_{q + 1} \rangle + \sum_{i = 0}^{q +1} \sum_{i< j}\langle a_0, \cdots, \hat a_{i}, \cdots, \hat a_{j}, \cdots, a_{q + 1} \rangle$$
ここで$\hat a_{i}$記号はその頂点を取り除くことを意味するものとする.$k < l$とすると上記の第一項において$i = l, j = l$としたものと第二項において$i = k, j = l$としたものは等しい.$k < l$は任意だったから
$$\sum_{i = 0}^{q + 1}\sum_{j< i}\langle a_0, \cdots, \hat a_{j}, \cdots, \hat a_{i}, \cdots, a_{q + 1} \rangle + \sum_{i = 0}^{q +1} \sum_{i< j}\langle a_0, \cdots, \hat a_{i}, \cdots, \hat a_{j}, \cdots, a_{q + 1} \rangle = \boldsymbol{0}$$
である.従って$\partial_{q} (\partial_{q + 1} (\sigma)) = \boldsymbol{0}$となる.$\square$

&emsp;命題1.2.3は集合の言葉で言い換えると
$$Im \partial_{q +1} \subset \ker \partial_q$$
を意味している.今までは$1$次のサイクルとバウンダリについてのみ考えていたが,上記の結果から一般の次元にも適用することが可能になる.

#### 定義
$z \in \ker \partial_q$を$q$**次元サイクル**といい,$b \in Im \partial_{q + 1}$を$q$**次元バウンダリ**という.

&emsp;以上のことから,「中身が詰まった図形」は$q$次元バウンダリとして厳密に表現できる.従って「穴になっているループのような図形」はサイクルの内バウンダリでないものとして表現できる.集合としては純粋に$\ker \partial_q - Im \partial_{q + 1}$としたいところだが,この集合は単体分割$K$の取り方に大きく変動してしまう.\
&emsp;そこで単体分割の取り方に依存せずに決まるものとして表現することを考える.まず,１つのサイクルにバウンダリを足してもまたサイクルであることは明らかである.従ってこのバウンダリの違いを無視した集合を考えたい.$Im \partial_{q +1} \subset \ker \partial_q$であったことと,この2つの集合はベクトル空間$C_q(K)$の部分ベクトル空間であることから商ベクトル空間の考えを適用できる.つまり次のように定義する.


#### 定義
$\ker \partial_q$上に同値関係$\sim$を定義する.
$$z \sim z' (z, z' \in \ker \partial_q) \overset{\text{def}}{\Longleftrightarrow} z - z' \in Im \partial_{q + 1}$$
この同値関係による商集合を$H_q(K)$と書く.$[z]$で$z \in \ker \partial_q$の同値類を表し,$z$の**ホモロジー類**,あるいは単に**$q$次元ホモロジー類**という.$H_q(K)$上には$\ker \partial_q$から次のような和とスカラー倍が矛盾なく定義される.
$$c[z] +c' [z'] := [cz + c'z'], \ \ \  c,c' \in \mathbb{Z}_2, z,z' \in \ker \partial_q$$
これにより$H_q(K)$は体$\mathbb{Z}_2$上のベクトル空間になる.この$H_q(K)$を**$q$次元ホモロジー群**という.

#### 注意
&emsp;一般にベクトル空間$V$とその部分空間$W$に対して,以下のような同値関係$\sim$
$$v \sim v' (v, v' \in V) \overset{\text{def}}{\Longleftrightarrow} v - v' \in W$$
を与えて得られる商集合は再びベクトル空間となり,これを商ベクトル空間と言って一般に$V/W$と書く.これに倣いホモロジー群も
$$H_q(K) = \ker \partial_q/Im \partial_{q + 1}$$
と表す.以降では$V/W$と書いたらそれは商ベクトル空間を表すものとする.

### ホモトピー同値

#### 定義
2つの位相空間$X,Y$の間の連続写像の族$\{f_t\}_{t \in [0, 1]}:X \longrightarrow Y$を考える.このとき,写像
$$H(s, t) = f_t(s)$$
が連続であるとき,これを$f_0$と$f_1$の間のホモトピーという.またこのとき$f_0$と$f_1$はホモトピックであるといい$f_0 \simeq f_1$と書く.

#### 定義
2つの位相空間$X,Y$の間に2つの連続写像$f: X \longrightarrow Y, g:Y \longrightarrow X$が存在して
$$g \circ f \simeq id_{X} \ \ \ f \circ g \simeq id_{Y}$$
となるとき,$X$と$Y$はホモトピー同値であるという.

&emsp;ホモロジー類は,直感には中身の詰まったサイクルを消して得られたものであるため,事実上同値関係$\sim$はホモトピー同値を代数的に表現したものといえる.つまり次の定理が成り立つ.
#### 定理
単体複体$K,L$に対し,多面体$|K|$と$|L|$がホモトピー同値なら$H_q(K) \simeq H_q(L)$

　この定理によってホモロジー群が単体分割の違いに依らずに定まることがわかる.

### 誘導準同型

&emsp; 次にパーシステントホモロジーの構成に必要な二つのホモロジー群の間の準同型性について述べる.二つの単体複体の間に単体写像がある場合,そこから自然にホモロジー群の間に準同型な対応を与える.これが誘導準同型写像と呼ばれる.

&emsp;誘導準同型写像を定義するにあたって鎖群の間の準同型写像である鎖準同型について定義する.
#### 定義
$K,L$を単体複体とし,$\partial_q:C_q(K) \longrightarrow C_{q - 1}(K), \partial '_q : C_q(L) \longrightarrow C_{q - 1}(L)$をそれぞれの境界準同型とする.写像の族$\{ \varphi_q : C_q(K) \longrightarrow C_q(L)\}$が
$$\forall q \in \mathbb{Z}, \varphi_{q - 1} \circ \partial_{q} =  \partial'_q \circ \varphi_q $$
を満たすとき,$\{ \varphi_q \}$を**鎖準同型写像**という.

#### 例
$\phi : K \longrightarrow L$を単体写像とする.$K$の$q$次元単体$\langle a_0, \cdots, a_{q+1} \rangle \in C_q(K)$に対して,準同型$\phi_q: C_q(K) \longrightarrow C_q(L)$を
$$\phi_q(\langle a_0, \cdots, a_{q+1} \rangle) := \begin{cases} \langle \phi(a_0), \cdots , \phi(a_q) \rangle && \text{if} \dim \phi ( \{ a_0, \cdots, a_q \}) =q \\ \boldsymbol{0} && \rm{else} \end{cases}$$
と定義すると$\{\phi_q\}$は鎖準同型写像となる.

&emsp;誘導準同型は鎖準同型を用いて次のような手順で定義される.

&emsp;まず$\varphi_q: C_q(K) \longrightarrow C_q(L)$を鎖準同型とする.$z \in Ker \ \partial_q$とするとき,$\partial_q'(\varphi_q(z)) = \varphi_{q - 1}(\partial_q(z)) = \varphi_{q - 1}(\boldsymbol{0}) = \boldsymbol{0}$となるから
$$z \in Ker \ \partial_q  \Longrightarrow \varphi_q(z) \in Ker \ \partial_q' $$
が任意の$q \in \mathbb{Z}$で成り立つ.\
&emsp;また$z,z' \in Ker \ \partial_q $のホモロジー類が同じ,つまり$z - z' \in Im \ \partial_{q + 1}$であるとする.この時$\varphi_q (z) - \varphi_q(z') \in \varphi_q(Im \ \partial_{q + 1})$となるが $\varphi_q(Im \ \partial_{q + 1}) = Im \ \varphi_q \circ \partial_{q + 1}  = Im \ \partial_{q + 1}' \circ \varphi_{q + 1}  = \partial_{q + 1}'(Im \ \varphi_{q + 1}) \subset Im \ \partial_{q + 1}'$ であるから
$$z - z' \in Im \ \partial_q \Longrightarrow \varphi_q(z) - \varphi_q(z') \in Im \ \partial_{q + 1}'$$
が任意の$q \in \mathbb{Z}$で成り立つ.\
&emsp;以上の結果から,ホモロジー群の間の準同型$\varphi_{q*}:H_q(K) \longrightarrow H_q(L)$が
$$\varphi_{q*} ([z]) = [\varphi_q(z)]$$
によって矛盾なく定義される.これを鎖準同型${\varphi_q}$による**誘導準同型写像**という.これを下の定義にまとめる.
#### 定義
ホモロジー群の間の準同型写像$\varphi_{q*}:H_q(K) \longrightarrow H_q(L)$を鎖準同型${\varphi_q}$による**誘導準同型写像**という.

### 相対ホモロジー
&emsp; 今までは単体複体のホモロジーについて一通り述べたが,一般に単体複体の中には部分複体の構造が複数ある.ある部分複体を無視したときに得られる単体複体のホモロジーについて記述したものが相対ホモロジー群である. 相対ホモロジー群は主に後の章において使用するためここでは定義のみを述べるにとどめておくことにする.

#### 定義

$K$を複体として$L \subset K$を部分複体とする.このとき鎖群$C_q(K),C_q(L)$は$C_q(L) \subset C_q(K)$を満たす.そこで$q$**次元相対鎖群**$C_q(K, L)$を 
$$C_q(K, L) := C_q(K) / C_q(L)$$
とおく.さらにこの定義において$L = \emptyset$の場合も許することにし,その際の相対鎖群は通常の鎖群$C_q(K)$を表すものとする.
$\partial_q: C_q(K) \longrightarrow C_{q - 1}(K)$を境界準同型とすると$\partial_q(C_q(L)) \subset C_{q - 1}(L)$であるから
$$c - c' \in C_q(L) (c , c' \in C_q(K)) \Longrightarrow \partial_q(c) - \partial(c') \in C_{q - 1}(L)$$
が成り立つ.従って相対鎖群の間の境界準同型$\partial_q^*:C_q(K, L) \longrightarrow C_{q - 1}(K, L)$が
$$\partial_q^*([c]) := [\partial_q(c)]$$
として矛盾なく定義される.\
$\partial_{q + 1} \circ \partial_q = \boldsymbol{0}$より$\partial_{q + 1}^* \circ \partial_q^*= \boldsymbol{0}$であるから,鎖群のときと同様に
$$Im \partial_{q + 1}^*\subset  \ker  \partial_q^* $$
が成り立つ.
$$H_q(K, L) : = \ker  \partial_q^*/Im \partial_{q + 1}^*$$
としたものを **$q$次元相対ホモロジー群** という.

## 多様体のホモロジー

### ホモロジー多様体

#### 定義
位相空間$M$が単体分割可能であり,その単体分割を$t:K \longrightarrow M$とし,$p \in M$とする.
$$H_q(L_M(p)) := H_q(L_K(t^{-1}(p)))$$
とおくとき,これを$M$の点$p$における**局所ホモロジー群**という.\
さらに,ある自然数$n$が存在して,任意の点$p \in M$に対して
$$H_q(L_M(p)) \simeq H_q(S^{n -1}) \ (q = 0, 1, \cdots)$$
が常に成り立つとき,$M$を **$n$次元ホモロジー多様体** という.

#### 例
円周$S^1$はホモロジー多様体である.

#### 定理
$n$次元ホモロジー多様体$M$の単体分割を$t:K \longrightarrow M$とすると,一般に次の性質が成り立つことが知られている.
$$(i) \dim K = n$$ 
$$(ii) \forall \sigma \in K \exists \tau \in K s.t. \sigma \prec \tau , \dim \tau = n$$
$$(iii) \sigma \in K \dim \sigma = n - 1 \Longrightarrow \# \{\tau \in K | \sigma \prec \tau, \dim \tau = n \} = 2$$

### 向きづけ可能性

&emsp;鎖群やホモロジー群の係数としては$\mathbb{Z}_2$を考えたが,$\mathbb{Z}$に置き換えても同様な方法でホモロジー群を構成できる.係数の集合を変更することによって多様体の向きづけ可能性を判定できるようになる.

#### 定義
有限集合$\{1, \cdots, q\}$からそれ自身への全単射$s:\{1, \cdots, q\} \longrightarrow \{1, \cdots, q\}$のことを$q$次の置換という.$q$次の置換全体を$S_q$で表すことにする.

#### 定義

$$S_q = A_q \bigsqcup B_q$$
とできる.このとき集合$\{A_q, B_q\}$の要素を **$q$次の向き**,あるいは単に**向き**という.また向き$s$とは異なる向きを$s$の**逆向き**といい,$-s$で表す.$1$次の置換は一つしかないから,向きはただ一つとしかない.

となるものとして定義する.ただしここで$sgn(s)$は$s$が偶置換の集合のときに$1$,奇置換の集合のときに$-1$となる数とする.この$s_i$を$\sigma$の$q-1$辺単体$\sigma_i = |a_0, \cdots,\hat a_i \cdots, a_q|$($\hat a_i$は$a_i$を取り除くの意味)に$s$から**導かれる向き**という.$s$から導かれる向きのついた$q - 1$辺単体$(\sigma_{i}, s_i)$を,$\hat \sigma_{s_i}$で表す

#### 定義

#### 定義
$n$次元ホモロジー多様体$M$の単体分割を$t:K \longrightarrow M$とすると,上述の定理1.1.2の性質$(iii)$から$K$の$n-1$単体はちょうど二つの$n$単体の辺単体となっている.$K$の全ての$n$単体に向きを定め,共通の$n-1$単体を持つ任意の二つの$n$単体の向きが同調しているようにできるとき,$n$次元ホモロジー多様体$M$は**向きづけ可能**であるという.

#### 定義
単体複体$K$の各$q$単体$\sigma_{i} \ i = 1, \cdots, m_q$に向きをそれぞれ定め,向きのついた単体$[\sigma_i] i = 1, \cdots, m_q$を作り,これらを基底とする自由$\mathbb{Z}$加群を$C_q(K, \mathbb{Z})$と書いて,$\mathbb{Z}$係数の$K$の$q$次元鎖群という.  
&emsp;$[\sigma_i]$の逆向き$-[\sigma_i]$と$C_q(K, \mathbb{Z})$元$(-1)[\sigma_i]$を同一視することで$C_q(K, \mathbb{Z})$は向きの取り方によらずに決まる.またこの同一視によって$sgn(s)=1$のときの$\sigma_s$をそのまま$\sigma$で書くと,$C_q(K, \mathbb{Z})$上では$\sigma_s = sgn(s)\sigma$が成り立つ.

#### 定義
$q$次元境界準同型$\partial_q:C_{q}(K,\mathbb{Z}) \longrightarrow C_{q}(K, \mathbb{Z})$を,$\sigma_s = |a_0, \cdots, a_q|_s$に対して
$$\partial_q(\sigma_s) = \sum_{i = 0}^{q}\hat \sigma_{s_i} = \sum_{i = 0}^{q}(-1)^i sgn(s) \sigma_i$$
で定義する.ここで$\hat \sigma_{s_i}$は$s$から$\sigma_i = |a_0, \cdots,\hat a_i \cdots, a_q|$に導かれる向き$s_i$を入れた$q -1$辺単体$(\sigma_i, s_i)$のことである.なお$\partial_0 = \boldsymbol{0}$と定義する.

#### 定理
上記の境界準同型に対して
$\partial_{q - 1} \circ \partial_{q} = \boldsymbol{0}$が成り立つ.

$proof)sgn(s) = 1$のときのみを考えればよい.
$$\partial_{q -1 } \circ \partial_q(\sigma) = \partial_{q - 1}(\sum_{i = 0}^{q}(-1)^i \sigma_i) =  \sum_{i = 0}^{q}(\sum_{j < i}(-1)^{i + j}(\sigma_i)_j + \sum_{ i < j}(-1)^{i + j - 1}(\sigma_i)_j)$$
ここで$k < l$とするとき,$i = k, j = l$のとき第二項と,$i = l, j = k$としたときの第一項が打ち消し合うから
$$ \sum_{i = 0}^{q}(\sum_{j < i}(-1)^{i + j}(\sigma_i)_j + \sum_{ i < j}(-1)^{i + j - 1}(\sigma_i)_j) = \boldsymbol{0}$$
が成り立つ.$\square$


&emsp;$\mathbb{Z}_2$係数のときと同様にサイクルとバウダリ,そしてホモロジー群を定義する.

#### 定義
$\text{Ker} \partial_q$の元を$q$次元サイクル,$\text{Im}\partial_{q + 1}$の元を$q$次元バウンダリという.$\text{Ker} \partial_q$も$\text{Im}\partial_{q + 1}$も$C_q(K, \mathbb{Z})$の部分加群であり,上記の定理より
$$\text{Im}\partial_{q + 1} \subset \text{Ker}\partial_q$$
が成り立つから,$\text{Ker} \partial_q$の同値関係$\sim$を
$$z \sim z' (z, z' \in \text{Ker} \partial_q) \Longleftrightarrow z - z' \in \text{Im}\partial_{q + 1}$$
によって定義できる.この商集合を$H_q(K, \mathbb{Z})$と書く.

#### 命題
$H_q(K, \mathbb{Z})$に次のような加法と整数倍が矛盾なく定義される.
$$[z] + [z'] := [z + z'] \\ n[z] := [nz]$$
この加法と整数倍により,$H_q(K,\mathbb{Z})$は$\mathbb{Z}$加群となる.$H_q(K,\mathbb{Z})$を$\mathbb{Z}$係数の$q$次元ホモロジー群という

&emsp;$\mathbb{Z}$係数のサイクルはどのような図形なのかについて考察する.定義式
$$\partial_q(\sigma_s) = \sum_{i = 0}^{q}(-1)^i sgn(\langle \sigma \rangle) \sigma_i$$
に注目する.各$\sigma_i$は当然$C_q(K, \mathbb{Z})$の基底になっている.従って,$\sigma_i$を辺単体とする他の$q$単体$\sigma'$が存在するなら$\partial_q(n \sigma + m \sigma') = 0 (n, m > 0)$となるためには,$\sigma_i$の係数が$0$すなわち$n(-1)^i sgn(\langle \sigma \rangle) + m(-1)^{i'}sgn(\langle \sigma' \rangle) = 0$でなければならない.この式は$|n| = |m|$となるが,$n, m >0$よりこれは$n = m$を意味する.よって
$$(-1)^i sgn(\langle \sigma \rangle) = - (-1)^{i'}sgn(\langle \sigma' \rangle)$$

#### 定理
弧状連結な$n$次元ホモロジー多様体$M$において次が成り立つ.\
$(i)H_n(M, \mathbb{Z}) = 0 \ or \  \mathbb{Z}$\
$(ii)M$が向きづけ可能である$\Longleftrightarrow H_n(M,\mathbb{Z})\simeq \mathbb{Z}$

弧状連結で向づけ可能な$n$次元ホモロジー多様体$M$に対して,$H_n(M,\mathbb{Z})$の生成元を一つ定めることを$M$に**向きをつける**と言い,この生成元を$M$の**基本ホモロジー類**と言う.  
&emsp;ちなみに$\mathbb{Z}_2$係数のホモロジー群$H_n(M)$では向きづけ可能性を判別することは不可能で,常に$H_n(M) \simeq \mathbb{Z}_2$である.それは$\mathbb{Z}_2$係数では$-1 = 1$が成り立つため,向きをどのように入れても自明に同調するからである.

# コホモロジー
&emsp;ホモロジーの双対の概念としてコホモロジーがある.コホモロジーは多様体の向きづけ可能性を議論する際によく使用される概念である.大雑把に言うと単体複体内の単体を線形(準同型)写像として捉え直したことで得られるベクトル空間である.
&emsp;本論文の目的であるホイットニー類のパーシステント化をする際にもやはりこのコホモロジーを使用する.しかしコホモロジーはホモロジーに比べてやや抽象度が高い.そこでこの章ではコホモロジーの定義について述べた後,それが前述したホモロジーとどのような関係性を持つのかについても記述する.

## コホモロジー群
&emsp;コホモロジー群の定義に入る前に幾つかの準備を行う.

#### 定義
$V$を$\mathbb{Z}_2$上のベクトル空間とし,$\hom(V, \mathbb{Z}_2 )$を$V$から$\mathbb{Z}_2$への線型写像全体の集合とする.$\hom(V, \mathbb{Z}_2 )$上に加法$f + g$とスカラー倍$cf$を
$$(f + g)(v) := f(v) + g(v) \\ (cf)(v) := cf(v)$$
で定義する.ただしここで$f,g \in \hom(V, \mathbb{Z}_2 ),c \in \mathbb{Z}_2, v \in V$としている.この加法とスカラー倍によって$\hom(V, \mathbb{Z}_2 )$は$\mathbb{Z}_2$上のベクトル空間になる.以降,$\hom(V, \mathbb{Z}_2 )$をベクトル空間$V$の**双対**と呼ぶことにする.

#### 命題
$m$次元ベクトル空間$V$の基底を$v_1, \cdots, v_m$として,これに対する$v_1^*, \cdots, v_m^* \in \hom (V, \mathbb{Z}_2)$を
$$v_i^* (v_j) = \delta_{ij}  \ \ \ (0 \leq i, j \leq m)$$
となる線型写像として定義する.ここで$\delta_{ij}$はクロネッカーのデルタである.このとき$v_1^*, \cdots, v_m^* $は$\hom (V, \mathbb{Z}_2)$の基底をなす.

$proof)$基底であることを示すには一次独立な生成系であることを示せば良い.\
&emsp;まずは$v_1^*, \cdots, v_m^* $が$\hom(V, \mathbb{Z}_2)$を生成することをみる.そのために$f \in \hom(V, \mathbb{Z}_2)$を任意とする.$v \in V$とすると,$v_1, \cdots, v_m$は$V$の基底であるから,ある$c_1, \cdots, c_m \in \mathbb{Z}_2$によって$v = c_1 v_1 + \cdots + c_m v_m$と書くことができる.$f$は線形であるから
$$f(v) = c_1f(v_1) + \cdots + c_m f(v_m)$$
となる.一方で$v_i^*$も線形であるから同様の式を満たすが,$v_i^*$の定義と合わせることで
$$v_i^*(v) = c_i$$
がわかる.$c^*_i:= f(v_i) \in \mathbb{Z}_2$とおくと,
$$f(v) = c_1^* v_1^*(v) + \cdots + c_m^* v_m^*(v)$$
を得る.$v \in V$は任意だったため,以上により$v_1^*, \cdots, v_m^* $は$\hom(V, \mathbb{Z}_2)$を生成することがわかる.\
&emsp;最後に$v_1^*, \cdots, v_m^* $が$\hom(V, \mathbb{Z}_2)$上一次独立であることを示す.そのために$c_1^*, \cdots, c_m^* \in \mathbb{Z}_2$が
$$\boldsymbol{0} = c_1^* v_i^*+ \cdots + c_m^* v_m^*$$
を満たすと仮定する.なおこの等式は写像としての等式であり,左辺の$\boldsymbol{0}$は$V$から$\boldsymbol{0} \in \mathbb{Z}_2$への定値写像を意味している.両辺の写像に各$v_i$における値を考えることで
$$ \boldsymbol{0} = c_i^*$$
を得る.従って$v_1^*, \cdots, v_m^* $が$\hom(V, \mathbb{Z}_2)$上一次独立である.\
&emsp;以上により$v_1^* \cdots, v_m^* $は$\hom(V, \mathbb{Z}_2)$の基底をなすことが示された.$\square$

#### 系
$\mathbb{Z}_2$上のベクトル空間$V$はその双対に線形同型である.$V \simeq \hom (V, \mathbb{Z}_2)$

#### 定義
命題2.1.2の基底$v_1^* \cdots, v_m^*$を,$V$の基底$v_1, \cdots, v_m$に対する**双対基底**という.

#### 定義
単体複体$K$に対する$q$次元鎖群$C_q(K)$の双対を$C^q(K)$と書いて,$K$の$q$**次元双対鎖群**という.さらに$\partial_{q + 1}:C_{q + 1}(K) \longrightarrow C_{q}(K)$を境界準同型とするとき,**$q$次元双対境界準同型** $\delta_q:C^q(K) \longrightarrow C^{q + 1}(K)$を
$$\delta_q(f) = f \circ \partial_{q + 1} \ \ (f \in C^q(K))$$ 
によって定義する.
$\partial_{q} \circ \partial_{q + 1} = \boldsymbol{0}$だから
$$\delta_{q} \circ \delta_{q - 1} = \boldsymbol{0}$$
が成り立つ.つまり$Z^q(K) = Ker \delta_q \ \ , B^q(K) = Im \delta_{q - 1}$とすると
$$B^q(K) \subset Z^q(K)$$
が成り立つ.\
そこで$q$**次元コホモロジー群**$H^q(K)$を以下で定義する.
$$H^q(K) := Z^q(K)/B^q(K)$$

&emsp;1章では相対ホモロジーを導入したが,同じようにコホモロジーに対しても相対コホモロジーを考えることができる.これはいわば相対ホモロジーの双対である.

#### 定義
相対鎖群$C_q(K, L)$から$\mathbb{Z_2}$への準同型写像全体の集合を$C^q(K, L)$と書く.$\partial_q^*:C_q(K, L) \longrightarrow C_{q - 1}(K, L)$を相対鎖群の境界準同型とするとき,その双対境界準同型$\delta_q^*:C^q(K, L) \longrightarrow C^{q + 1}(K, L)$を
$$\delta_q^*(f) : = f \circ \partial_{q + 1}^* \ \ \ (f \in C^q(K, L))$$
で定義する.$\partial_q^*\circ \partial_{q + 1}^* = \boldsymbol{0}$であるから,$\delta_{q} ^* \circ \delta_{q - 1}^*= \boldsymbol{0}$である.つまり
$$Im \ \delta_{q - 1}^* \subset \ker  \delta_{q}^*$$
が成り立つ.そこで相対コホモロジー群$H^q(K, L)$を
$$H^q(K, L) := \ker  \delta_{q}^*/ Im \delta_{q - 1}^*$$
で定義する.

&emsp;次にこのコホモロジー群とホモロジー群の関係性について述べている二つの定理について記述する.後々の議論のため定理内ではホモロジーもコホモロジーも相対になっているが,$L = \emptyset$とすることで通常のホモロジーとコホモロジーの間にも成り立つ定理である.

#### 定理
(普遍係数定理)
相対コホモロジー群$H^q(K, L)$は相対ホモロジー群の双対$\hom(H_q(K, L), \mathbb{Z}_2)$と同型である.

$proof)$二つの線形写像$\phi:H^q(K, L) \longrightarrow \hom(H_q(K, L), \mathbb{Z}_2)$と$\psi:\hom(H_q(K, L), \mathbb{Z}_2)\longrightarrow H^q(K, L)$を構成して$\psi \circ \phi$と$\phi \circ \psi$が共に恒等写像であることを示す.

- $\phi$の構成\
    コホモロジー類$[f] \in H^q(K, L)$に対して$\phi([f]) \in \hom(H_q(K, L), \mathbb{Z}_2)$を次で定義する.
    $$\phi([f])([z]) :=  f(z)$$
    この$\phi([f])$は見かけ上代表元の取り方に依存している.$\phi([f])$がホモロジー群から$\mathbb{Z}_2$への写像として矛盾なく定義されるには,
    $$f \sim g , z \sim z' \ (f, g \in \ker  \delta_{q}^*, z, z' \in \ker  \partial_q^*) \Longrightarrow f(z) = g(z')$$
    となることを確認する必要がある.ところがこれは,それぞれの同値関係の定義から$f - g = \delta_{q - 1}^* (h), z - z' = \partial_{q + 1}^*(c) \ (h \in C^{q - 1} (K, L), c \in C_{q + 1}(K, L))$とおくと
    $$f(z) - g(z') = (f(z) - g(z)) +( g(z) - g(z')) \\ = (f - g)(z) + g(z - z') \\ = \delta_{q - 1}^* (h)(z) + g(\partial_{q + 1}^*(c)) \\ = h(\partial_q^*(z)) + \delta_q^*(g)(c) \\ = \boldsymbol{0}$$
    が得られることから明らかである.従って$\phi([f])$は$f$の同値類によってのみ決定される写像$\phi([f]):H_q(K, L) \longrightarrow \mathbb{Z}_2$であり,定義から明らかに$H_q(K, L)$上線形であるから$\phi([f]) \in \hom(H_q(K, L), \mathbb{Z}_2)$が確かにわかる.また
    $$\phi([f] + [g]) = \phi([f]) + \phi([g]) \\ \phi(c[f]) = c\phi([f]) \ \ (c \in \mathbb{Z}_2)$$
    であることも$\phi([f])$の定義から確認できる.\
    以上によって線型写像$\phi:H^q(K, L) \longrightarrow \hom(H_q(K, L), \mathbb{Z}_2)$が構成された.

- $\psi$の構成\
    $C_q(K, L)$はベクトル空間で$\partial_q^*:C_q(K, L) \longrightarrow C_{q - 1}(K, L)$は線形写像であるから,線形代数における一般論より,ある部分空間$W$が存在して
    $$C_q(K, L) = \ker  \partial_q^* \oplus W$$
    と直和分解され$\eta := \partial^*_q | _W : W \longrightarrow Im \partial^*_q$が同型写像となる.そこで$F \in \hom(H_q(K, L), \mathbb{Z}_2)$に対して,写像$h_F:C_q(K, L) \longrightarrow \mathbb{Z}_2$を,$c = z + w \in C_q(K, L) \ (z \in \ker  \partial_q^*, w \in W)$に対して
    $$ h_F(c) := F([z])$$
    と定義する.$h_F$は$c \in C_q(K, L)$に関して線形であるから$h_F \in C^q(K, L)$であり,特に$c \in C_{q + 1}(K, L)$とすると
    $$\delta_q^*(h_F)(c) = h_F (\partial_{q +1}^*(c)) = F([\partial_{q +1}^*(c)]) = F(\boldsymbol{0}) = \boldsymbol{0}$$
    となるから$\delta_q^*(h_F) = \boldsymbol{0}$つまり$h_F \in \ker \delta_q^*$となる.そこで目的の写像$\psi:\hom(H_q(K, L), \mathbb{Z}_2)\longrightarrow H^q(K, L)$を
    $$\psi(F) := [h_F]$$
    で定義する.$h_F$の定義から明らかに$\psi$は$F \in \hom(H_q(K, L), \mathbb{Z}_2)$に関して準同型である.

- $\psi \circ \phi$が恒等写像であること\
    $[f] \in H^q(K, L)$とする.$\psi \circ \phi ([f]) = [h_{\phi([f])}]$に注意すると,$f \sim h_{\phi([f])}$を示せば良い.$c = z + w \in C_q(K, L) \ (z \in \ker  \partial_q^*, w \in W)$とすると
    $$f(c) - h_{\phi([f])}(c) = f(c) - \phi([f])([z])  \\  = f(c) - f(z) = f(c - z)  = f(w)   \\ = f \circ \eta^{-1} \circ \eta(w)  = f \circ \eta^{-1}(\partial_q^*(w)) \ \ (\because \eta(w) = \partial_q^*(w)) \\ = f \circ \eta^{-1}(\partial_q^*(c)) \ \ (\because \partial_q^*(c) = \partial_q^*(w)) \\ = f^*(\partial_q^*(c)) \ \ (\because f^*|_{Im \partial_q^*} = f \circ \eta^{-1}) \\ =\delta_{q - 1}^*(f^*)(c)$$
    従って$f - h_{\phi([f])} = \delta_q^*(f^*)$となる.ただし,ここで線形写像$f^*:C_{q - 1}(K, L) \longrightarrow \mathbb{Z}_2$は,線形写像$f \circ \eta^{-1}:Im \partial_q^* \longrightarrow \mathbb{Z}_2$をベクトル空間の直和分解$C_q(K, L) = Im \partial_q^*\oplus V$に対して,$f^*|_{Im \partial_q^*} = f \circ \eta^{-1}, f^*(V) = \boldsymbol{0}$と拡張したものとしている.以上の式変形によって$f \sim h_{\phi([f])}$が成り立つ.

- $\phi \circ \psi$が恒等写像であること\
    $F \in \hom(H_q(K, L), \mathbb{Z}_2)$とする.$\phi \circ \psi(F) = \phi([h_F])$より$\phi([h_F]) = F$を示せば良い.$[z] \in H_q(K, L)$に対して$\phi([h_F])([z]) = h_F(z) = F([z])$だから確かに$\phi([h_F]) = F$が成り立つ.

以上の議論から題意が示された.$\square$

#### 系
$\mathbb{Z}_2$係数においてはホモロジーとコホモロジーは同型である.
$$H_k(K) \simeq H^k(K)$$

#### 定義
$f:H_k(L) \longrightarrow H_k(K)$を準同型とする.普遍係数定理により$H^k(K) \simeq \hom (H_k(K), \mathbb{Z}_2)$とみなすことで,各コホモロジー類$r \in H^k(K)$は線形写像$r: H_{k}(K) \longrightarrow \mathbb{Z}_2$として扱える.従って準同型$f^*:H^k(K) \longrightarrow H^k(L)$が次のように定義される.
$$f^*(r) :=r \circ f \ \ ( r \in H_k(K))$$
この$f^*$を**$f$に対するコホモロジーの誘導準同型**という.

## 積

#### 定義
$p$次元双対鎖$c \in C^p(K )$と$q$次元双対鎖$d \in C^q(K )$に対して$c \smile d \in C^{p + q}(K )$を得る操作を**カップ積**と言い,以下で定義する.
$$(c \smile d)(\langle a_0, \cdots, a_{p + q} \rangle) := c(\langle a_0, \cdots , a_p \rangle) \cdot d(\langle a_{p}, \cdots , a_{p+q} \rangle )$$
このとき
$$\delta_{p + q}(c \smile d) = (\delta_{p} c) \smile d + c \smile \delta_q d$$
を満たすからこれによってコホモロジー類同士のカップ積$H^p(K ) \times H^q(K ) \longrightarrow H^{p + q}(K )$が矛盾なく定義される.
$$\lbrack z \rbrack \smile \lbrack z' \rbrack =\lbrack z \smile  z' \rbrack$$

#### 定義
$p + q$次元鎖$c \in C_{p+q}(K )$と$f \in C^p(K )$に対して,その**キャップ積**$c \frown f \in C_q(K )$を以下で定義する.
$$ g(c \frown f) := (f \smile g)(c)  \ \ , g \in C^q(K )$$
キャップ積は以下の性質を持つ.
$$\partial (c \frown f) = \{(\partial c \frown f) - c \frown \delta f\} $$ 
これによってホモロジー類とコホモロジー類のカップ積$H_{p + q}(K ) \times H^p(K ) \longrightarrow H_{q}(K )$が矛盾なく定義される.
$$\lbrack z \rbrack \frown \lbrack f \rbrack =\lbrack z \frown f \rbrack$$

加群の係数の環として$\mathbb{Z}$を用いたが,$\mathbb{Z}/2\mathbb{Z}$としても同様の議論が可能である.以下では$\mathbb{Z}/2\mathbb{Z}$を係数とする場合は$C_*(K \mathbb{Z}/2)$などと表す.ホモロジー群や双対鎖群,コホモロジー群についても同様である.

#### 定理
弧状連結で向づけ可能な$n$次元ホモロジー多様体$M$に対し
$$H^q(M) \simeq H_{n - q}(M) \ \ (q = 0, 1, \cdots)$$
であり,その同型写像$l_q:H^q(M) \simeq H_{n - q}(M)$は定義2.2.4のキャップ積と$M$の基本ホモロジー類$[M]$を用いて
$$l_q([z]) = [M] \frown [f]$$
で与えられる.

##  ホイットニー類

#### 定義
$E, B$を位相空間として,$\pi: E \longrightarrow B$を連続な全射とする.今,任意の$b_0 \in B$に対して次の$(i),(ii)$を満たすとき,$\xi = (E, B,\pi)$をベクトル束という.\
$(i) F_{b_0}(\xi ) := \pi^{-1} (b_0)$は体$\mathbb{R}$上のベクトル空間である.\
$(ii) b_0$における$B$のある開近傍$U \subset B$と同相写像$h:U \times \mathbb{R}^n \longrightarrow \pi^{-1}(U)$が存在して,対応
$$\mathbb{R}^n \ni x \longmapsto h(b, x) \in F_b(\xi)$$
が各$b \in U$においてベクトル空間としての同型を与える.
$E$をベクトル束の全空間,$B$を底空間,$\pi$を射影といい$E(\xi), B(\xi), \pi_\xi$などと書く.また$(ii)$の$(U,h)$を$b$のまわりの$\xi$に対する局所座標系という.


#### 例
多様体$M$の接束$TM$はベクトル束である.

#### 例
実射影空間$P^n = \{\{x, -x \}| x \in S^n \}$を底空間とする**直線束**と呼ばれるベクトル束$\gamma _n^1 = (E, P^n, \pi)$は次のように定義される.
$$F_{\{x, -x\}}(\gamma_n^1)= \{(\{x, -x\}, \lambda x)|  \lambda \in \mathbb{R} \}\\ E = \bigcup_{b \in P^n}F_b(\gamma_n^1) \\ \pi(F_b(\gamma_n^1)) = b$$

#### 定義
全空間の間の連続写像$f:E(\xi) \longrightarrow E(\eta)$が,各$F_b(\xi)$に対してある$F_{b'}(\eta)$と線型同型を与えるとき,$f$を**束写像**という.$\hat f(b) = b'$とすると$f$は連続写像$\hat f: B(\xi) \longrightarrow B(\eta)$を与える.



#### 定義
$\xi = (E, B, \pi)$をベクトル束とする.底空間の部分集合$B' \subset B$を取ると,$B'$上に$E' := \pi^{-1}(B')$を全空間とし$\pi' := \pi|_{E'}$を射影とするベクトル束$(E', B', \pi')$が構成される.これを$\xi|{B'}$と書いて$\xi$の$B'$への**制限**という.


#### 定義
$\xi = (E, B, \pi)$をベクトル束とし,勝手な位相空間$B'$から$B$への連続写像を$f:B' \longrightarrow B$とする.このとき,$B'$を底空間とするベクトル束$f^*\xi = (E', B', \pi')$が次のようにして構成できる:
$$E' := \{(b, e) \in B' \times E|f(b) = \pi(e) \} \\ \pi'(b, e) := b \ \ \ \ \ (b, e) \in E'$$
ただしここでファイバー$F_b(f^* \xi)$上のベクトル空間としての演算は
$$\lambda(b, e_1) + \mu(b, e_2) :=  (b, \lambda e_1 + \mu e_2) \ \ \ (\lambda , \mu  \in \mathbb{R})$$
と定義する.こうしてできたベクトル束$f^*\xi$を**誘導束**という.なお,定義2.3.1の$(ii)$(局所自明性という）は以下のようにして確認できる:\
$b_0 \in B'$とし$(U, h)$を$f(b_0)$のまわりの$\xi$に対する局所座標系とする.$U' := f^{-1}(U)$とするとこれは$b_0$の$B'$における開近傍である.そこで
$$h': U' \times \mathbb{R}^n \longrightarrow \pi'^{-1}(U')$$
を$h'(b, x) := (b, h(f(b), x))$で定義すると$(U', h')$が求める$b_0$の局所座標系となる.


#### 定義
$\xi_1 = (E_1, B_1, \pi_1),\xi_2 = (E_1, B_2, \pi_2)$をベクトル束とする.このとき$E_1 \times E_2$を全空間,$B_1 \times B_2$を底空間,$\pi_1 \times \pi_2$を射影とするベクトル束$\xi_1 \times \xi_2$が構成できる.これを$\xi_1$と$\xi_2$の**直積束**という.


#### 定義
底空間が同じ$B$である二つのベクトル束を$\xi, \eta$とする.$d(b) := (b, b)$で定義される対角線写像$d:B \longrightarrow B \times B$は連続である.よって$B$上にベクトル束$d^*(\xi \times \eta)$が構成されるが,これを$\xi$と$\eta$のホイットニー和といい$\xi \oplus \eta$と書く.なおホイットニー和の各ファイバー$F_b(\xi \oplus \eta)$は標準的に$F_b(\xi) \oplus F_b(\eta)$に同型である.


#### 定義
ベクトル束$\xi$に対し,以下の4つの公理を満たすコホモロジー類を**ホイットニー類**という.

$(i)w_i(\xi) \in H^i(B(\xi)) (w_0(\xi) = 1,\ i = 0, 1, \cdots )$

$(ii)$束写像から定まる連続写像$f:B(\xi)  \longrightarrow B(\eta)$に対して$w_i(\xi) = w_i(f^*\eta)$

$(iii)\xi, \eta \ (B(\xi) = B(\eta)) \ \Longrightarrow \ w_k(\xi \oplus \eta) = \sum_{i} w_i(\xi)w_{k - i}(\eta)$

$(iv)$円周$P^1$上の直線束$\gamma_1^1$に対して$w_i(\gamma_1^{1}) \ne \boldsymbol{0}$
ホイットニー類は一意的に存在する.

# パーシステント$\cdot $(コ)ホモロジー
&emsp;あるパラメータ(例えば時間)によって単体複体が増大していく際,その多面体内の穴が,ある時刻で生まれてさらに進んだ時刻で消えるといった現象が観察される.パーシステントホモロジーはそういった現象をホモロジー群の生成元によって記録するものである.\
&emsp;しかし,通常のパーシステントホモロジーは向き付け可能性の可否を判別することはできていない.本論文ではこの問題を解消するために,パーシステント性をコホモロジーのレベルに発展させたパーシステントコホモロジーおよびパーシステントホイットニー類を構成する.


## 構造定理

#### 定理
$F = \oplus_{i = 1}^{m}( x^d_i)$の斉次部分加群$H \subset G \subset F$の階数を$n, m$とする.このとき$s \in \mathbb{Z}_{\geq 0}$と自然数$l_1 \leq \cdots, \leq l_s$が存在して剰余加群$G / H$は$G$のある斉次基底$g_1, \cdots, g_m$を用いて
$$G / H = \bigoplus_{i = 1}^{s} \langle [g_i] \rangle \oplus  \bigoplus_{i = n + 1}^{m} \langle [g_i]  \\ \text{Ann}([g_i]) = (x^{l_i}), \ i = 1, \cdots, s \\ \text{Ann}([g_i]) = 0, \ i = n + 1, \cdots, m$$
と表せる.ここで$s = 0$のときは$\text{Ann}([g_i]) \ne 0 $となる成分は表れない.なお$\text{Ann}(m) := \{f \in \mathbb{Z}_2[x]| f \cdot m = 0 \}$は$m$の**零化イデアル**という　

#### 定理
$M = \oplus_{i \geq 0} M_i$を有限生成次数付き$\mathbb{Z}_2[x]$加群とする.このとき$M$は次数付き$\mathbb{Z}_2[x]$加群としての同型
$$M \simeq \bigoplus_{i = 0}^{s} \left( (x^{d_i}) / (x^{d_i + l_i}) \right ) \oplus \bigoplus_{i = s + 1}^{s + r}(x^{d_i})$$
の形に表せる.さらに$d_i, l_i, s, r$は$M$の次数付きの同型類から一意に決まる.

## パーシステントホモロジー

#### 定義
集合列$\{A_i\}_{i = 0}^{\infty}$が単調増加であるとき,つまり
$$A_0 \subset A_1 \subset \cdots $$
が成り立つとき,$\{A_i\}_{i = 0}^{\infty}$をフィルトレーションという..フィルトレーションにおいては添字$i$は**時刻**と呼ばれる. またある非負整数$\theta$が存在して$i \geq \theta $なら$A_i = A_{\theta}$が成り立つとき,フィルトレーション$\mathbb{A}$は**有限型**であるといい,そのような$\theta$の内で最小のものを**飽和時刻**という. この論文では有限型のみを扱う.    
&emsp;以降でフィルトレーションを表す際は$\mathbb{A}$や$\mathbb{X}$のような二重線の白抜き文字を用いる.あるいは単調増加な関係性を明確にするために
$$\mathbb{A}: A_0 \subset A_1 \subset \cdots $$
のようにして表すこともある.仮にフィルトレーションを単に$\mathbb{A}$や$\mathbb{X}$と書いた場合フィルトレーションの$i$番目の要素(集合)を,$A_i$や$X_i$のように二重線を無くした通常の表記に添字をつけた記号で表すことにする.   

#### 定義
単体複体のフィルトレーション$\mathbb{K}: K_0 \subset K_1 \subset \cdots $を考え,$K = \bigcup_{i \geq 0} K_i$とおく.$\sigma \in K$が時刻$t$で発生したとは$\sigma \in K_t -  K_{ t - 1}$を満たすことであり,このとき$T(\sigma) = t$と表す.

#### 例
ヴィートリスリップス複体$VR_r(P)$からはパラメータ$r$に関するフィルトレーションが得られる.つまり$r < r'$なら$VR_r(P) \subset VR_{r'}(P)$が成り立つ.
$P$は有限集合であるからこのフィルトレーションは有限型であり,有限個の対$(r, r')$を除いて$VR_r(P) =  VR_{r'}(P)$が成り立つ.したがってパラメータの列$0 = r_0< r_1< \cdots < r_\Theta$に対して有限型のフィルトレーション
$$VR_{r_0}(P) \subset VR_{r_1}(P) \subset \cdots VR_{r_{\Theta}}(P )$$
を得る.

#### 例
ヴィートリスリップス複体の時と同様に$0 = r_0< r_1< \cdots < r_\Theta$に対してチェック複体の有限型のフィルトレーション
$$C_{r_0}(P) \subset C_{r_1}(P) \subset \cdots C_{r_{\Theta}}(P )$$
を得る.脈体定理によりチェック複体$C_r(P)$は$X_r = \bigcup_{x \in P}B(x, r)$にホモトピー同値である.

### パーシステントホイットニー類
ここではホイットニー類のパーシステント性について定義する。ホイットニー類はベクトル束上で定義される概念であったが,これをパーシステントコホモロジーにおける概念に拡張するために幾つかの準備を行う.

#### 定義
$E$を全空間,$X$を底空間とするベクトル束$\xi = (E,X, \pi)$を考える.今飽和時刻$\Theta$において$X_{\Theta} = X$となるフィルトレーション
$$\mathbb{X}:\emptyset \subset X_0 \subset X_1 \subset \cdots \subset X_{\Theta} = X$$

## 自然画像データのパーシステントホモロジー

&emsp;　パーシステントホモロジーやホイットニー類の応用先として,Mumfordによる自然の背景画像の特徴に関する研究がある.一般に画像は,グレースケールのデータとして扱う場合,数学的には高次元のベクトル空間$P$のベクトルとして扱うことができる.従って自然の背景画像の集合も高次元ベクトル空間のある部分集合$T$と見做せる.このことから,写真のデータセットがある多様体を近似していることを想定した研究が今までにいくつか行われてきた.しかしMumfordは別の手法をとった.それはデータセット内の各画像から$3 \times 3$のパッチをランダムにとり,そのパッチの空間を分析する方法である.このような手法を取った理由は画像の空間の次元にある.仮に自然の背景画像の集合がある多様体を近似していたとしても,$P$から任意に取ってきた画像が自然の画像である可能性は低く$T$の次元も余次元もかなり高いことが予想され,結果的に空間全体の分析を行うのが難しくなるのである.そこでMumfordは代わりに$3 \times 3$のパッチの空間の局所的な特徴に関する研究を行った.その結果,$3 \times 3$のパッチが成す空間には高密度な2次元部分多様体が存在する十分な証拠が得られた. それに対してCarlssonは同じようなパッチの空間とその部分集合に対して,多項式の空間によるパラメータ化を行い,結果としてKlein bottleと同相な2次元の部分集合がパッチの空間内に存在することを証明した.以下ではその概要を示す.

# 多様体上の点データに対するパーシステントホモロジー

&emsp;Carlssonたちは自然画像によるパッチの空間内にKlein bottleが存在することを証明した.一般に,ユークリッド空間内のあるデータの集合がある低次元多様体を近似していると考える仮説を**多様体仮説**というが,近似する多様体が$n$次元ホモロジー多様体であるなら,向きづけ可能なものとそうでないものに分類でき,$n$次のホモロジーは$\mathbb{Z}$係数では二種類しかない.そこで近似する多様体を求めるという視点ではなく,すでに知られている多様体上にデータの集合が散らばっていると考え,多様体を切り替えるごとに得られる特徴量をデータの特徴とみなす方法を考える.  
&emsp;似たような考えに基づいた統計学として,角度(方向)統計学がある.例えば地球上のある地点における風向きの観測データは角度のデータとして扱われるのだが,角度は$2\pi$の周期で同一視される特異性から通常の実数直線を標本空間として想定すると期待値や分散が不自然なものになる.これに対して方向統計学では角度データを統計的に正しく扱うために角度を円周上の$1$点と捉えた.つまり標本空間を$S^1$と捉えて期待値や分散を定義している.  
&emsp;角度データのように多様体の上にあると想定されたデータ集合上で脈体に基づく複体を用いたパーシステントホモロジーの計算を行う際も,ユークリッド空間上の点として通常の距離を用いるのは不自然と考えられる.そこでユークリッド空間上の点集合を多様体の点集合として捉える方法を述べる.

## 多様体ごとの距離

&emsp;一般にリーマン多様体は距離空間としてみなすことができ,ナッシュの埋め込み定理によってすべてのリーマン多様体はあるユークリッド空間内へ等長に埋め込むことができる.したがって点データの集合に想定する多様体の位相と一致する距離を定義すれば,想定する多様体を得られたことになる.そこで以降はデータの集合と想定する多様体ごとに適切な距離を導入する.  
&emsp;なお,想定する$n$次元多様体$M$に対して,データ集合$D$が$D \subset \mathbb{R}^N( N > n)$なら,適当な次元削減$h:\mathbb{R}^N \longrightarrow \mathbb{R}^n $を行うことで,$h(D) \subset \mathbb{R}^k$をデータの集合として捉えることにする.そして,$h(D)$に$n$次元多様体$M$から定まる距離を計算し,$h(D)$を頂点とし,距離を重みとする重み付きグラフ$G(D)$を考える.この重み付きグラフから単体複体$\Delta(D)$を構成する.

#### $M = S^1$のとき

&emsp;まず$\mathbb{R}$から$S^1\subset \mathbb{R}^2$への連続写像$f:\mathbb{R} \longrightarrow S^1$として$f(x) = (\cos 2 \pi x,  \sin 2 \pi x)$を考える.
その上で$\mathbb{R}$上に同値関係$\sim$を以下のように定めた商位相空間$\mathbb{R}/ \sim$を考える.
$$a \sim b \overset{\text{def}}{\Longleftrightarrow} a - b \in \mathbb{Z}$$
<!-- ここで$\pi : \mathbb{R} \longrightarrow \mathbb{R}/ \sim$を標準的射影とする.同値関係の定義から$\forall x \in \mathbb{R}$と$ 0 < \forall \epsilon < 1 / 2$に対して
$$\pi^{-1}(\pi((x - \epsilon , x + \epsilon))) = \bigcup_{n \in \mathbb{Z}} (x + n  - \epsilon, x + n + \epsilon)$$
が成り立つ.右辺は$\mathbb{R}$の開集合であるから$\pi((x - \epsilon , x + \epsilon))$は商空間$\mathbb{R}/\mathbb{Z}$の開集合となる.従って$\pi$は開写像となるから -->
ここで
$$a \sim b \Longleftrightarrow f(a) = f(b)$$
が成り立つ.従って連続写像
$$\hat f : \mathbb{R}/\sim \ni [x] \longmapsto f(x)  \in S^1$$
が矛盾なく定義される.$\hat f$の逆写像として
$$g : S^1 \ni  (\cos \theta, \sin \theta) ( 0 \leq \theta < 2\pi) \longmapsto  [\frac{\theta}{2 \pi}]  \in \mathbb{R}/ \sim $$
が考えられるがこれもまた連続であることから$\hat f$は同相写像となる.そこで$\mathbb{R} \times \mathbb{R}$上の非負関数$d(x, y)$として次のように定義する.
$$d_{S^1}(x, y) := \|f(x) - f(y) \|$$
このとき
$$d_{S^1}(x, y) = 0 \Longleftrightarrow x \sim y \\  x \sim x', y \sim y' \Longrightarrow d_{S^1}(x, y) = d_{S^1}(x', y') \\ d_{S^1}(x, y) = d_{S^1}(y, x) \\ d_{S^1}(x, y) \leq d_{S^1}(x, z) + d_{S^1}(z, y)$$
となるから$d_{S^1}$は$\mathbb{R} / \sim$の商位相の距離関数となる.距離をこれで導入しても問題ないが,計算のしやすさという観点から,もう少し簡単な式に置き換えることにする.
$d_{S^1}$は$d_{S^1}^2(x,y) = 2 - 2\cos 2 \pi |x - y|$となるから距離を計算する際は代わりに$1 - \cos 2\pi|x - y|$を計算してもよい.この関数は$z = |x - y|$とするとき,$0 \leq z \leq 1/2$で単調増加,$1/2 \leq z \leq 1$で単調減少,$z \in \{0, 1\} $のとき$0$になる関数であるから,距離の大小だけに注目するときは代わりに$1  - |2z - 1| $を計算すればよい.またこれは$2 \min(z, 1 -z)$と一致するため,最終的に
$$d_{S^1}(x, y) = \min(|x -y|, 1 -|x - y|)$$
として計算すればよい.

#### $M = P^1$のとき
$f:\mathbb{R} \longrightarrow S^1$を前節のものと同じとする.その上で$\mathbb{R}$上に同値関係$\sim '$を以下で定める.
$$x \sim' y (x, y \in \mathbb{R}) \Longleftrightarrow f(x)  = f(y) \ \text{or} \ f(x) = - f(y)$$
このとき
$$x \sim' y  \Longleftrightarrow \ 2x - 2y \in \mathbb{Z}$$
が成り立つ. そこで
$$d_{P^1}(x, y) := \| f(2x) - f(2y) \|$$
と定義すれば,これが商位相空間$\mathbb{R}/ \sim'$上の距離関数となる.$x \in \mathbb{R}$の同値な$y \in \mathbb{R}$は$y = x + n / 2 \ ( n \in \mathbb{Z})$の形で表されるから
$$f(x + n/2) = (\cos (2\pi x + n\pi), \sin (2 \pi x + n \pi)) = \pm (\cos 2\pi x, \sin 2 \pi x) = \pm f(x)$$
となるから写像
$$\hat f: \mathbb{R}/\sim \ni [x] \longmapsto \{f(x), -f(x)\} \in P^{-1}$$
が矛盾なく定義され,とくに同相写像となる.なお,$P^1$と$S^1$は位相的には同相となる.

### $n$次元トーラス$T^n = S^1 \times \cdots \times S^1$

&emsp;各データ$p \in \mathbb{R}^n$を$n$次元トーラス$T^n$に埋め込む方法を考える. $T^n$は直積距離空間として捉えることができるため,$S^1$への埋め込みを用いて,$x = (x_1, \cdots, x_n), y = (y_1, \cdots, y_n) \in \mathbb{R}^n$に対して$d_{T^n}(x, y) := \sum_{i = 1}^{n} d_{S^1}(x_i,y_i)$として定義すればよい.

### Klein bottle
&emsp;$I = [-1, 1]$の直積$I \times I$上のデータ$p \in I \times I$同士をKlein bottleの距離で計算する方法を与える.$I \times I$の点$p_1$はその属する象限に応じて$\mathbb{R}^2$内で下図のような同値な点をもつ.

## 角度統計学

角度に値をとる確率変数に対する統計学が近年研究されている.それは角度統計学や方向統計学と呼ばれている.角度データは通常のデータと異なり,$2\pi$の周期で一致するという特殊性がある.そのような背景から通常の統計量との整合性が取れず,適切なものを定義する必要がある.
ここでは角度統計学で一般的な確率分布や統計量について簡単に記述する.なお,角度の範囲としては主に$[0, 2\pi), [- \pi, \pi)$などが用いられるがこの区間のいずれかを一括で表すために$\Pi$という記号を用いることにする.

有限個の角度データの標本を$\theta_1 , \cdots, \theta_n \in \Pi$とする.

### 標本平均方向
**標本平均方向**$\bar \theta$を
$$\bar \theta = \arg(\sum_{j = 1}^{n}e^{i \theta_j})$$
で定義する.ただし,$\sum_{j = 1}^{n}e^{i \theta_j}$が$\boldsymbol{0}$ベクトルの時は平均方向は定義されない.
$C = \sum_{i = 1}^n \cos \theta_i, S = \sum_{i = 1}^n \sin \theta_i, R = \sqrt{C^2 + S^2}$とするとき,標本平均方向$\bar \theta$は,$\Pi = [0, 2 \pi)$のときは次のように陽の形で解くことができる.
$$\bar \theta = \begin{cases} \tan^{-1} (S/C) & C > 0, S \geq 0 \\ \pi/2 & C=0, S>0 \\ \tan^{-1}(S/C) + \pi & C < 0 \\ 3\pi / 2 & C=0, S<0  \\ \tan^{-1}(S/C) + 2\pi & C > 0, S<0\end{cases}$$

### 標本平均合成ベクトル長(mean resultant length)と角度分散(circlar variance)
標本平均合成ベクトル長$\bar R$を
$$\bar R = \frac{1}{n} | \sum_{j = 1}^{n} e^{i \theta_j} |$$
で定義する.$0 \leq \bar R \leq 1$であり,$\bar R = 1$となるときは各$\theta_j$が一致することと必要十分であることが次のようにしてわかる.
$$\begin{split}\bar R = 1 &\iff C^2 + S^2 = n^2\\ &\iff n + 2 \sum_{1 \leq i < j \leq n}(\cos \theta_i \cos \theta_j + \sin \theta_i \sin \theta_j) = n^2\\ &\iff \sum_{1 \leq i < j \leq n}\cos(\theta_i - \theta_j) = n(n - 1)/2 \\ &\iff \cos(\theta_i - \theta_j) = 1 \ \ \ 1 \leq i < j \leq n  \\ &\iff \theta_i = \theta_j \ \ \ 1 \leq i < j \leq n \end{split}$$
したがって$\bar R$は標本の集中度を表していると考えられる.そこで
$$\nu = 1 - \bar R$$
を標本角度分散として定義する.標本角度分散は$0 \leq \nu \leq 1$を満たしている.これは円周上の点がどれだけ離れていても対照点にしかなり得ないことを示している.

### 平均方向と角度分散の性質
標本平均は偏差の和が$0$になるという性質がある.標本平均方向も似たような性質を持つ.$\bar \theta$は
$$\sum_{j=1}^n \sin(\theta_j - \bar \theta) = 0$$
を満たすことが簡単な計算でわかる.また,単位円周上の2つの角度$\theta, \alpha$に対応する2点の間の弦の長さは余弦定理によって$2(1 - \cos (\theta - \alpha))$と書くことができるため,角度$\alpha$からの差の2乗の平均は
$$V(\alpha) = 1 - n^{-1} \sum_{j = 1}^{n} \cos(\theta_j - \alpha)$$
と表され,これは$V(\alpha) = \nu + 2 \bar R \sin^2(\frac{\bar \theta - \alpha}{2})$と変形できることから,$V(\alpha)$は$\alpha = \bar \theta$のときに最小であり$V(\bar \theta) = \nu$が最小値となることがわかる.

### 標本$p$次三角モーメント
$p \in \mathbb{Z}$に対して$\bar C_p = \frac{1}{n} \sum_{j = 1}^n \cos p\theta_j, \bar S_p = \frac{1}{n} \sum_{j = 1}^n \sin p\theta_j$とするとき
$$m'_p = \frac{1}{n}\sum_{j = 1}^n e^{ip \theta_j} = \bar C_p + i \bar S_p$$
を標本$p$次三角モーメントという.$m'_p = \hat \rho_{p}e^{i \hat \mu'_{p}} \ \ (\hat \rho_p \geq 0)$のとき
$$\hat \rho_p^2 = \bar C_p^2 + \bar S_p^2, \ \ \ \cos \hat \mu'_{p} = \frac{\bar C_p}{\hat \rho_p}, \ \ \ \sin \hat \mu'_{p} = \frac{\bar S_p}{\hat \rho_p}$$
がなりたつ.$\hat \rho_{p}$は標本$p$次平均合成ベクトル長,$\hat \mu'_{p}$は標本$p$次平均方向という.$p = 1$のときはそれぞれ通常の標本平均合成ベクトル長や標本平均方向に一致する.

### 平均方向周りの標本$p$次三角モーメント
標本平均方向$\bar \theta$周りの$p$次三角モーメントを
$$m_p = \frac{1}{n} \sum_{i = 1}^n \cos \{p(\theta_j - \bar \theta)\} + i \frac{1}{n} \sum_{i = 1}^n \sin \{p(\theta_j - \bar \theta)\}$$
で定義する.$m_p$の長さは$\hat \rho_p$に等しく,$m_p$の偏角は$\arg m_p = \hat \mu'_p - p \bar \theta$となる.

### 円周上の確率密度関数と分布関数
円周上の確率分布の確率密度関数とは次の性質を持つ関数$f(\theta)$のことを言う.$[0, 2\pi)$とする.

$(1)$ $0 \leq \theta < 2 \pi$に対して,$f(\theta) \geq 0$\
$(2)$ $\int_{0}^{2\pi} f(\theta) d\theta = 1$\
$(3)$ $f(\theta)$は周期$2\pi$を持つ周期関数である.つまり$\forall k \in \mathbb{Z}$に対して,$f(\theta + 2 k \pi) = f(\theta)$

$(2)$と$(3)$より,$f(\theta)$は長さが$2\pi$の任意の区間$[\theta_L, \theta_U)$に対して$\int_{\theta_L}^{\theta_U} f(\theta) d\theta = 1$が成り立つ.


$$F(\theta )=\left\{2\pi I_{0}(\beta )\right\}^{-1}\left[\theta I_{0}(\beta )+2\left\{\sum _{j=0}^{\infty }{\frac {I_{j}(\beta )\sin(j(\theta -\mu ))}{j}}\right\}\right]$$

$$f(\theta )={\frac {\exp\{\beta \cos(\theta -\mu )\}}{2\pi I_{0}(\beta )}}$$

$$I_{j}(\beta )=\left({\frac {\beta }{2}}\right)^{j}\sum _{i=0}^{\infty }{\frac {\left({\frac {\beta ^{2}}{4}}\right)^{i}}{i!\Gamma (j+i+1)}}$$
